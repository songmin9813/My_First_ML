```python
#X y 
X = np.array(data.iloc[:, data.columns != 'Class'])
y = np.array(data.iloc[:, data.columns == 'Class'])
print("Shape of X: {}".format(X.shape))
print("Shape of y: {}".format(y.shape))
```

```python
#X,y  ,   
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
def model_evaluation(label, predict):
    cf_matrix = confusion_matrix(label, predict)
    Accuracy = (cf_matrix[0][0] + cf_matrix[1][1]) / sum(sum(cf_matrix))
    Precision = cf_matrix[1][1] / (cf_matrix[1][1] + cf_matrix[0][1])
    Recall = cf_matrix[1][1] / (cf_matrix[1][1] + cf_matrix[1][0])
    F1_Score = (2 * Recall * Precision) / (Recall + Precision)
    print("Model_Evaluation with Label:1")
    print("Accuracy: ", Accuracy)
    print("Precision: ", Precision)
    print("Recall: ", Recall)
    print("F1-Score: ", F1_Score)
```

# LightBGM   

```python
from sklearn.metrics import confusion_matrix
import lightgbm as lgb
lgb_dtrain = lgb.Dataset(data = pd.DataFrame(X_train), label = pd.DataFrame(y_train)) #   LightGBM   
lgb_param = {'max_depth': 10, #  
            'learning_rate': 0.01, # Step Size
            'n_estimators': 50, # Number of trees,   
            'objective': 'multiclass', #  .    (0 1 )
            'num_class': len(set(pd.DataFrame(y_train))) + 1} #  , Label must be in [0, num_class) -> num_class 1 .
lgb_model = lgb.train(params = lgb_param, train_set = lgb_dtrain) #  
lgb_model_predict = np.argmax(lgb_model.predict(X_test), axis = 1) #   , Softmax      Label 
model_evaluation(y_test, lgb_model_predict) #    
```

# SMOTE    

```python
#  X_train, y_train, X_test, y_test  
print("Number transactions X_train dataset: ", X_train.shape)
print("Number transactions y_train dataset: ", y_train.shape)
print("Number transactions X_test dataset: ", X_test.shape)
print("Number transactions y_test dataset: ", y_test.shape)
```

```python
from imblearn.over_sampling import SMOTE
print("Before OverSampling, counts of label '1': {}".format(sum(y_train == 1))) # y_train    1  
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train == 0))) # y_train    0   

sm = SMOTE(random_state = 42, sampling_strategy = 0.3) # SMOTE ,  . 0  30  
X_train_res, y_train_res = sm.fit_resample(X_train, y_train.ravel()) # Over Sampling 

print("After OverSampling, counts of label '1': {}".format(sum(y_train_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train_res==0)))
```

```python
print("Before OverSampling, the shape of X_train: {}".format(X_train.shape)) # SMOTE    
print("Before OverSampling, the shape of y_train: {}".format(y_train.shape)) # SMOTE    
print('After OverSampling, the shape of X_train: {}'.format(X_train_res.shape)) # SMOTE   
print('After OverSampling, the shape of y_train: {}'.format(y_train_res.shape)) # # SMOTE   
```

```python
lgb_dtrain2 = lgb.Dataset(data = pd.DataFrame(X_train_res), label = pd.DataFrame(y_train_res)) #   LightGBM   
lgb_param2 = {'max_depth': 10, #  
            'learning_rate': 0.01, # Step Size
            'n_estimators': 50, # Number of trees,   
            'objective': 'multiclass', #  
            'num_class': len(set(pd.DataFrame(y_train_res))) + 1} #  , Label must be in [0, num_class) -> num_class 1 .
lgb_model2 = lgb.train(params = lgb_param2, train_set = lgb_dtrain2) #  
lgb_model2_predict = np.argmax(lgb_model2.predict(X_test), axis = 1) #   , Softmax      Label 
model_evaluation(y_test, lgb_model2_predict) #    
```
