## ML의 종류-supervised learning
- 선형 회귀 분석(Linear Regression)
  - 독립변수와 종속변수가 선형적인 관계가 있다라는 가정하에 분석
  - 직선을 통해 종속변수(Y)를 예측하기 때문에 독립변수(X)의 중요도와 영향력을 파악하기 쉬움
  - 올라갔다 내려가는 모델의 경우 선형이 아님에 유의(비선형)

- 의사 결정 나무(Decision Tree)
  - 독립 변수의 조건에 따라 종속변수를 분리(비가 내린다 -> 축구를 하지 않는다.)
  - 이해하기 쉬우나 overfitting이 잘 일어남
    - 앙상블 기법과 혼용하서 잘 사용함

- KNN(K-Nearest Neighbor)
  - 새로 들어온 데이터의 주변 k개의 데이터의 class로 분류하는 기법
  - Hyper parameter : 사람이 지정해주어야 하는 변수를 의미

- Neural Network
  - 입력, 은닉, 출력층으로 구성된 모형으로써, 각 층을 연결하는 노드의 가중치를 업데이트하면서 학습
  - 딥러닝의 일종
  - overfitting이 잘 일어난다는 단점이 존재

- SVM(Support Vector Machine)
  - Class간의 거리(margin)가 최대가 되도록 decision boundary를 만드는 방법
  - 왼쪽, 오른쪽 클래스 사이 거리가 최대가 되는 경계선을 그리는 법
    - 이를 통해 특정 독립변수(X)에 대한 종속변수(Y)를 기대할 수 있음
  - 학습하는 시간이 오래 걸린다(다른 모델에 비해) 고로 최근에 잘 쓰이지 않는다.

- Ensemble Learning
  - 여러 개의 모델(classifier or base learner)을 결합하여 사용하는 모델
  - 아래의 그림은 여러 개의 의사 결정 트리를 만들어 투표를 통해 최종 클래스를 산출
  - base learner를 다양하게 만들 수 있을까가 앙상블 러닝의 핵심
  - 요즘 많이 사용한다고 하는 알고리즘이라 한다.

## ML의 종류-unsupervised learning
- K-means clustering
  - Label 없이 데이터의 군집으로 k개로 생성
  - 랜덤으로 두 점을 찍고 그 점의 중심점을 찍음. 중심점의 데이터들을 할당함을 반복하면서 군집화 형성
  - K에 따라서 성능이 많이 달라진다. 해당 데이터는 K=2일 경우에 가장 이상적인 데이터
