- 복합한 모형일수록, 데이터가 적을수록 과적합이 일어나기 쉽다.
- 아래 그림은 회귀분석에서 고차항을 넣었을 때 만들어지는 직선
- 과적합은 data science뿐만 아니라 AI 전반적으로 매우 큰 이슈
- 완화시키는 기술만 발전되어가고 있음

## 분산(Variance)과 편파성(Bias)의 트레이드오프(Tradeoff)(Dilemma) / 반비례 관계
  - 모형 f(X)로 모집단의 전체 데이터를 예측할 때 발생하는 총 error를 계산하면 reducible error와 irreducible error로 표현됨
  - 그리고 reducible error는 다시 분산과 편파성으로 구성된다.

> $E(Y-\hat{Y})^2=E[f(X)+\epsilon-\hat{f}(X)]^2 = Var(\hat{f}(X))+[Bias(\hat{f}(x)]^2+Var(\epsilon)$

- $E(Y-\hat{Y})^2$ : 오류자승의 평균(기대값) : mean squared error(MSE)
- $Var(\hat{f}(X))$ : 추정 모델의 분산
  - 전체 데이터 집합 중 다른 학습 데이터를 이용했을 때, $\hat{f}$이 변하는 정도
  - 복잡한 모형일수록 분산이 높다.
- $[Bias(\hat{f}(x)]^2$ : 실제 모델과의 편파성(얼마나 차이가 나는가)
  - 학습 알고리즘에서 잘못된 가정을 했을 때 발생하는 오차
  - 간단한 모형일수록 편파성이 높다.
- $Var(\epsilon)$ : irreducible error(우리가 찾아낼 수 없는 값의 분산값)

### 분산(Variance)와 편파성(Bias)의 트레이드 오프(Tradeoff)를 그림으로 표현

- Low Variance + Low Bias가 가장 이상적인 형태이나, 실 추정 모형은 Low Variance + HIgh Bias / HIgh Variance + Low Bias 형태
  - Low Variance + HIgh Bias : 단순한 모델을 만든 형태
  - HIgh Variance + Low Bias : 복잡한 모델을 만든 형태
- 따라서 적절한 모형 선택과 실험 설계를 통해 과적합을 방지한다.
- **앙상블 형태(HIgh Variance + Low Bias)에서 평균을 내는 것이 트렌드**

#### 학습 데이터가 전체 집단(모집단)의 모든 데이터를 보장할 수 없기 때문에 Overfitting에 유의하여야 한다.
