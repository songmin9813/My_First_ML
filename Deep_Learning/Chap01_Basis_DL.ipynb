## 1. 딥러닝 vs 기계학습 vs 인공지능
- 인공지능 : 기계가 사람의 행동을 모방하게 만드는 기술
- 기계 학습 : 기계가 일일이 코드로 명시하지 않은 동작을 데이터로부터 학습하여 실행할 수 있도록 하는 알고리즘을 개발하는 연구분야
- 딥러닝 : 기계 학습의 한 분야인 인공 신경망에 기반하여, 많은 양의 데이터를 학습해 뛰어난 성능을 이끌어내는 분야
- 인공지능이라는 큰 틀 아래 기계학습. 그 아래에 딥러닝이 있다고 생각하면 됨

## 인공지능과 딥러닝의 차이
- 머신러닝의 경우
  - 사진으로부터 특징 추출기를 이용하여 특징을 구분할 수 있는 벡터를 계산하여 분류기를 제작
- 딥러닝의 경우
  - Feature Extractor & Classifier를 사용하여 사진을 구분함


- 딥러닝으로 할 수 있는 것에는 분류와 회귀, 그리고 **물체 검출**(Object Detection)이 가능하다.
  - 물체 검출로는 one-shot detection이 각광을 받고 있음
- 외에도 영상 분할(Image Segmentation)과 영상 초해상도(Image Super Resolution)가 가능하다.
  - 차선 검출/영역 당 행동 등을 분석이 가능. 또는 해상도의 복원이 가능
- 예술적 창조물(GAN을 활용한 창조물)과 강화 학습 또한 딥러닝을 통하여 가능

## 딥러닝의 구성요소
- 앞으로도 계속 볼 거니 지금 다 알고 있을 필요는 없음
- 학습단계, 테스트 단계, 손실 함수 등이 있음 정도만 알고 있자

## 2. Deep Learning의 시작과 현재 


- 최초의 인공신경망 개념으로 나온 것은 McCulloch와 Pitts 교수에 의해 정의됨(1943년)
- Rosenblatt의 퍼셉트론 구조를 통해 입력을 통한 가중치. 퍼셉트론의 개념이 나옴(1958년)
  - 퍼셉트론의 수식과 시각적 표현
  - $y=+1$ 이라면 $b+w_1x_1+w_2x_2 \geq 0$
  - $y=-1$ 이라면 $b+w_1x_1+w_2x_2 < 0$


- 1 또는 -1이라는 하드한 표현을 사용하여 문제를 해결하는 방식으로 이해

## Perceptron과 ADALINE(Adaptive Linear Element)과의 비교


### XOR 문제와 도래된 AI Winter
- 그럼 XOR 문제는 어떻게 해결할 수 있지?
- 1969년 Perceptron(Minskey and Papert)에 의하여 문제 제기
- 이 문제를 해결할 수 없었기에 1970~1990년 동안 침체기가 생김

- **하나의 직선으로는 XOR을 온전히 표현할 수 없음!**

### 첫번째 breakthrough
- 다층 퍼셉트론을 이용하여 이를 해결하려는 노력
  - 중간에 은닉계층을 둔다면 직선을 두 개 그리는 효과를 줄 수 있다!
  

### 두 번째 breakthrough
- 오류 역전파 알고리즘
  - 정방향의 입력이 들어왔을 때 입력에서 바로 출력으로 가는 것이 아니라, **출력의 역방향으로 error를 한 번 계산하는 것이 더 효율적이라는 결론에 이르게 됨**
  

### 두 번째 AI Winter의 대두
- 계층이 깊어질수록(은닉 계층의 개수가 많아질수록) 학습이 어려운 기울기 소실이 발생함

### 이에 대한 breakthrough - 심층 믿음 신경망의 등장
- 비지도 학습이기 때문에 어떤 모델을 사용해도 유의미한 데이터를 추출해낼 수 있음
