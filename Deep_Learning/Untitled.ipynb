## 딥러닝 vs 기계학습 vs 인공지능
- 인공지능 : 기계가 사람의 행동을 모방하게 만드는 기술
- 기계 학습 : 기계가 일일이 코드로 명시하지 않은 동작을 데이터로부터 학습하여 실행할 수 있도록 하는 알고리즘을 개발하는 연구분야
- 딥러닝 : 기계 학습의 한 분야인 인공 신경망에 기반하여, 많은 양의 데이터를 학습해 뛰어난 성능을 이끌어내는 분야
- 인공지능이라는 큰 틀 아래 기계학습. 그 아래에 딥러닝이 있다고 생각하면 됨

## 인공지능과 딥러닝의 차이
- 머신러닝의 경우
  - 사진으로부터 특징 추출기를 이용하여 특징을 구분할 수 있는 벡터를 계산하여 분류기를 제작
- 딥러닝의 경우
  - Feature Extractor & Classifier를 사용하여 사진을 구분함
