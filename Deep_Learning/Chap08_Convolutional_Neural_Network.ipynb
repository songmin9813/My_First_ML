## 1. 아날로그 신호처리
### 선형 시불변 시스템(Linear Time Invariant System : LTI System)
- LTI 시스템은 선형적이고 시간에 영향을 받지 않는 신호처리 시스템
- a와 b라는 시스템, 그리고 시스템 f(x)관계가 있다고 할 때
  - $f(a+b)+f(a)+f(b)$관계가 성립하면 이는 선형관계에 있다고 한다.
- 언제 넣어도 동일한 값이 나오는 것

## Dirac 델타 함수

- $t=0$을 제외한 모든 위치에서 출력이 0
- 모든 구간에서 적분한 값이 1
  - h를 무한히 크게 한다 하여도 값은 1이 될 것

## 임펄스 응답

- LTI 시스템에 임펄스(Dirac 델타 함수)를 입력했을 때의 출력을 **임펄스 응답**이라고 한다.
- 임펄스 응답을 **필터**라고도 하며, LTI 시스템의 동작을 완전하게 표현한다.
  - LTI 시스템이 아닌 경우에는 이 관계가 성립하지 않음에 유의

## 합성곱 연산(Convolution)
- 두 함수를 합성하는 합성곱 연산
- **한 함수를 뒤집고 이동하면서, 두 함수의 곱을 적분하여 계산**

### LTI 시스템은 입력 신호에 임펄스 응답을 합성곱한 결과와 같다.


## 이차원 신호와 흑백 이미지

- 흑백 영상은 각 픽셀이 0~1 사이의 실수로 된 2-D Signal로 표현할 수 있다.

## 컬러 이미지

- 컬러 영상은 RGB의 3채널로 구성된 2-D Signal로 표현할 수 있다.

## 영상의 합성곱 계산

- 2-D 디지털 신호의 합성곱은 필터를 한 칸씩 옮기면서 영상과 겹치는 부분을 **모두 곱해 합치면 된다.**

## Padding이 필요한 이유

- 합성곱 연산 시, 필터의 크기에 따라 영상의 크기가 줄어드는 문제가 있다.
- 영상 크기 보존을 위해서는 Padding이 필요할 것!

## Zero-Padding

- 크기가 (2N+1)인 커널(필터)에 대해, 상하좌우에 N개의 Zero-Padding을 해주면 된다.
  - 원래 영상 크기보다 큰 크기의 필터를 만들어 영상과 같은 크기로 출력이 되도록 조정하는 것

## Stride : 크기를 줄이고 싶은 경우

- 합성곱 연산에서 커널을 이동시키는 거리를 Stride라고 하며, 이를 크게 하면 출력의 크기가 줄어든다.

## 잡음 제거 필터

- 2-D Gaussia Filter를 적용하면 흐려진 영상을 얻을 수 있다.
- 영상이 흐려지는 대신 **잡음을 제거하는 특성**이 있다.
  - 영상의 전반적인 내용 또한 확인할 수 있는 특징이 있음

## 미분 필터

- Sobel Filter를 적용하면 특정 방향으로 미분한 영상을 얻을 수 있다.
- 기울기가 빠르게 변하는 부분=해당 방향의 Edge 성분을 추출하는 특성을 가짐

## 합성곱 계층

- 익숙한 뉴런. 다시 한 번 상기하고 넘어갑시다.

## 2. 지금까지 배웠던 것을 하나로 묶어보자

- 입력 뉴런 대신 **입력 영상**을, 가중치 대신 **필터**를, 곱 대신 **합성곱**을, **편향은 그대로 유지하면**아름다운 신경망 형태가 나오게 된다.
  - scalar 형태로 존재하는 bias를 matrix를 tensorflow에서는 자동으로 변환하는 기능을 내장하고 있음

## Fully-Connected Layer

- FC 계층은 여러 개의 뉴런을 한 곳에 모아둔 것으로 Matrix의 곱연산 형태로 표현됨

## 2. 합성곱 계층

- 합성곱으로 이루어진 뉴런을 전결합 형태로 연결한 것을 **합성곱 계층**이라고 한다.
- 합성곱 계층은 주로 우측에 있는 방식으로 쉽게 표현한다.
  - 각각의 연결당 하나의 커널이 존재하기 때문에 $k_H \times k_W \times C_{in} \times C_{out}$과 같은 표현을 사용함
- 왼쪽 그림 : 각 행렬로 이루어진 C(입력)
- 오른쪽 그림 : 각 행렬로 이루어진 C(출력)
- 사이에 껴져있는 kernel 또한 height와 width로 표현함


- 합성곱 계층은 $C_{in} \times C_{out}$ 번의 **합성곱 연산**으로 이루어져 있다.
- 편향은 전결합 계층과 동일하게 하나의 벡터로 이루어진다.

## 합성곱 계층의 의미

- 여러 채널에서 특별한 특징이 나타나는 위치를 찾아내는 것이 합성곱 계층의 임무
- ex에서는 첫 합성곱 연산에서 Edge와 관련된 Feature를 추출하는 Layer를 보여주고 있음

## 3. 풀링 계층의 기원 : LeNet-5

- 1998년도로 상당히 이른 시기에 CNN의 기본적인 구조를 잘 정립하였다.
- 32에서 28 by 28로 줄어든 이유 : zero-padding을 진행하지 않은 커널이기 때문에 이는 5 by 5의 커널을 이용하였음을 알 수 있다.

## 3. 합성곱 신경망의 기본 구조

- 영상을 subsampling 해주는 계층 : 풀링 계층
- 평탄화 작업(Flatten)은 전결합 계층을 이용하여 진행하여야 함
  - 출력 계층의 개수를 맞추어주기 위함

## 합성곱 계층의 특징 : 상기하고 넘어가

- 합성곱 계층에서는 영상의 크기는 그대로이며, **영상의 채널 수가 달라진다.**
  - 물론 zero-padding의 여부, stride 여부에 따라 크기가 달라질 수 있다.
- 합성곱 계층에 의해 추출된 결과는 공간적 특징이 있으며 '**특징 맵**(Feature Map)'이라고 한다.

## 풀링 계층(Pooling Layer)

- 여러 화소를 종합하여 하나의 화소로 변환하는 계층
- 풀링 계층을 통과하면 **영상의 크기가 줄어들고, 정보가 종합되는 특징을 가진다.**

## 풀링 계층의 매개변수

- 풀링 계층은 Stride와 Kernel size를 변경하여 사용할 수 있다.
- 일반적으로 Stride=2, Kernel size=(2,2)를 사용한다.(Defalut)
  - 커널 사이즈는 출력 영상 크기에 영향을 주지 못함에 유의
  - 각 계층이 겹치지 않도록 설정하는 것이 일반적

## 다양한 풀링 방법

- **Max Pooling** : 각 커널 사이즈 안에서 가장 큰 수치를 종합.
  - 특징 맵에서 값이 높은 것이 중요할 때 사용하는 방법
- **Average Pooling** : 커널 안에서 평균을 내어 종합
  - 데이터들이 버려지지 않는 특징. 세세한 데이터까지 필요한 경우에 사용하는 방법
- 풀링 방법은 다양하지만, 가장 많이 쓰이는 방법은 **최댓값과 평균값**이다.
- 합성곱 신경마의 애플리케이션에 맞는 풀링 계층을 사용한다.

## Receptive Field

- 같은 크기의 필터여도, 풀링에 의해 작아진 특징 맵에 적용되면 원본 영상에서 차지하는 범위가 넓다.
  - 풀링 자체가 압출된 형태이기 때문에 하나의 커널이 넓은 영역을 커버칠 수 있는 효과가 있음
- 이 범위를 **Receptive Field**라고 한다.
