##    : XOR  

- NAND  OR   XOR    .

```python
#import module
import tensorflow as tf
import numpy as np
```

```python
#  
class Perceptron:
    def __init__(self,w,b):
        self.w=tf.Variable(w, dtype=tf.float32) #   
        self.b=tf.Variable(b, dtype=tf.float32) #  
        
    def __call__(self, x):
        return tf.sign(tf.reduce_sum(self.w*x)+self.b) #       
```

- sign    : -1 1   

```python
#   .  
def v(*args):
    return np.array(args) #  
```

```python
#  weight, bias 
w=v(1,1)
b=0.5

perceptron=Perceptron(w,b)
```

```python
#   
p1=perceptron(v(1,1)) #1. T T
p2=perceptron(v(-1,1)) #2. F T
p3=perceptron(v(-1,-1)) #3. F F
p4=perceptron(v(1,-1)) #4. T F

print(p2.numpy(), p1.numpy())
print(p3.numpy(), p4.numpy())
```

##  2 : XOR  

```python
# XOR  
#  1 : nand  
p_nand=Perceptron(w=v(-1,-1),b=0.5)
#  2 : or  
p_or=Perceptron(w=v(1,1),b=0.5)
#  3 :    and  
p_and=Perceptron(w=v(1,1),b=-0.5)

def xor(x):
    h1=p_nand(x)
    h2=p_or(x)
    return p_and(v(h1,h2))
```

```python
p1=xor(v(1,1)) #1. T T
p2=xor(v(-1,1)) #2. F T
p3=xor(v(-1,-1)) #3. F F
p4=xor(v(1,-1)) #4. T F

print(p2.numpy(), p1.numpy())
print(p3.numpy(), p4.numpy())
```
