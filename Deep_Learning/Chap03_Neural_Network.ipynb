## 퍼셉트론의 수학적 표현

- 퍼셉트론은 수학적으로 내적과 sign 함수로 간단하게 표현할 수 있다

  - 각 값을 곱하여 더하는 것이 내적의 기본 작동 방식
  - 각각의 퍼셉트론을 더한 값으로 표현된다는 개념이 내적의 개념과 같음

## Sigmoid Function

- 본래 S-Shape의 함수 모두를 가리키나, 딥러닝에서는 보통 위의 함수로 정의된다.
- Tanh / Soft Decision에 비해 **0과 1 사이의 값을 가지기 때문에 확률을 표현할 수 있다는 특징을 가짐**
  - T의 값을 구하게 된다면 자연스럽게 F의 값은 1-T이기 때문에 유용하게 사용됨

## Softmax Function

- multi class classification
- 여러 개의 입력을 받았을 때 각 값의 exp값을 취해준 함수
- 최종 출력 단계에서 **N가지 범주로 분류하는 multi-class classification에 쓰임**
- 입력값은 어느 값이나 들어올 수 있음 : $e^x$그래프의 특성상 모든 값이 0보다 큰 양수가 나오게 된다.
  - 이 모두를 더하여 나누어주게 되면 모든 값은 0보다 큰값이 되고, 더한 값은 1이 된다는 특징을 가지게 된다.

## Softmax vs Sigmoid
- Sigmoid는 **하나의 입력을 0으로 강제한 2-class softmax 함수와 동일**하다.
  - 2가지 클래스를 구분하기 위해 1개의 입력을 받는다는 점에 주목
  

## 번외 : ReLU Function - 다음에 또 나옴

- ReLU는 현재 쓰임을 자세히 알긴 어려우나 지속적으로 다루게 됨
- **0보다 작은 값을 강제로 0으로 만드는 함수**

## 손실 함수(Loss Function)
- 지도 학습 알고리즘에서 반드시 정의되어야 함
- 얼마나 못하는지를 표현하는 것.
- 최적화 이론에서 최소화하고자 하는 함수
- 알고리즘 학습 중 아직 **얼마나 못 하는지**에 대한 표현
- 보통 미분 가능한 함수를 사용
  - 최적화 방법에 미분을 사용한 방법이 많이 존재하기 때문
- 목적 함수, 비용 함수, 에너지 함수 등으로 불림
- 학습 과정에서 **중요한 역할을 함**

## 성능 척도(Performance measure)
- 학습된 알고리즘의 성능을 측정하는 지표
- 정량적으로 알고리즘을 비교/평가하기 위함
- 미분 가능 여부에 관계없이 사용
- 학습 과정에서는 사용하지 않음
- **비즈니스 목표**와 연관이 높음

## 알고리즘 학습의 수학적 의미

- 만약 arg min이 아닌 max가 적혀있다면 뒤의 목적함수(L)를 최대로 만드는 세타 값을 출력하는 것을 의미함
- 모델을 가지고 학습을 한다 : 여러 파라미터를 가지고 종합적인 학습을 한다는 것
  - 이 모든 것을 세타라는 값으로 하나로 모으는 작업을 학습이라고 함
  - 이 최적의 세타 값을 찾는 것이 중요한 것
- 오른쪽의 그림은 세타 값이 하나만을 가지고 있는 경우에 해당하는 그래프.
- **손실 함수는 작으면 작을수록 좋은 모델임을 알려주는 지표이기 때문에 반드시 필요하다 볼 수 있다.**

## 평균제곱오차(Mean squared error : MSE)

- 에러를 제곱하여 평균을 낸 것
- 가장 기본적인 손실 함수. 오차가 커질수록 손실함수가 빠르게 증가하는 특징이 있다. 회귀 분석에서 많이 쓰이는 지표
- 오른쪽의 그림은 linear한 상태가 아니기 때문에 입력값의 크기에 따라 그 출력값이 빠르게 변화하는 특징을 가짐
- 통계적으로 **평균값과 밀접한 관련이 있음**

## 평균절대오차(Mean Absolute Error : MAE)

- 오차가 커져도 손실함수가 일정하게 증가하는 특징이 있음. 이상치에 강건하다는 특징(Robust)을 가짐
- 통계적으로 **중간값(Median)과 연관이 있다.** 회귀 분석에 많이 쓰임
- 절대값을 사용하기 때문에 손심함수가 빠르게 증가하는 단점이 보완됨

## 원 핫 인코딩

- 원-핫 인코딩은 다중 클래스 분류 문제의 정답을 표기할 때 사용
- 정답 벡터의 각 인덱스 별로 클래스를 미리 정의하고, 정답에 해당하는 요소만 1로 표현
  - 이러한 정답 벡터를 Label이라고 부름

## 교차 엔트로피 오차(Cross entropy error : CEE)

- 원핫 인코딩으로 인해, 정답인 클래스에 대해서한 오차를 계산
- 정확히 맞추면 오차가 0, 틀릴수록 오차가 무한히 증가하는 특징을 가짐
- 값이 0과 1 사이의 값을 가짐에 유의

## 예제

- 정답인 1 부분을 제외하고 나머지 0 부분은 다 죽어버린다. 실제 정답에 해당하는 0.6, 0.8의 값에 대해서만 CEE 과정을 거침
- 오차를 내는 과정에서는 정답 클래스만 비교하지만, 다중 클래스 분류의 활성함수인 Softmax로 인해 다른 클래스에 대한 학습에도 영향을 준다.

## 신경망 구조 - Neuron

- 신경망은 **뉴런을 기본 단위로 하며** 이를 조합해 복잡한 구조를 이룬다.

## 신경망의 간결한 표현

- 보통 신경망을 표현할 때, 그래프의 노드와 엣지를 이용해 표현한다.
- 여기서 **노드는 단일 뉴런 연산을, 엣지는 뉴런의 연결성을 의미**한다.
- 각 벡터의 원소를 곱한 것을 $w\cdot x$로 표현할 수 있다.(벡터의 내적)

## Fully-Connected Layer(Dense Layer) : 전결합 계층
- 전결합 계층 : 두 계층 간의 모든 뉴런이 연결되어 있는 계층

- Tensorflow 에서는 **Dense**라고 표현되어있음
- w : 가중치
- x : 입력
- b : 편향

## Shallow Neural Network : 얕은 신경망

- 입력 계층(Input layer), 은닉 계층(Hidden layer), 출력 계층(Output layer)으로 이루어진 신경망
- 심층 신경망(Deep neural network : DNN)의 등장 이후 기존의 신경망을 얕은 신경망이라 부르기로 함
- 각 계층은 *Fully-connected layer*로 이루어져 있음
- 은닉 계층으로는 Tanh, Sigmoid 함수를 주로 사용하고, 출력계층으로는 Linear(단순 선형), Softmax, Sigmoid(바이너리 한정)을 주로 사용한다.

## Deep Neural Network : DNN(심층 신경망)

- 얕은 신경망보다 은닉 계층이 많은 신경망을 DNN이라고 부른다.
- 보통 5개 이상의 계층이 있는 경우 깊다라는 표현을 사용한다.
- DNN에서는 ReLU 함수를 많이 사용한다.

## 실습 시작 전 : XOR 게이트 구현하기

- NAND 게이트와 OR 게이트를 쌓아올려 XOR 게이트를 구현해낼 수 있다.

```python
#import module
import tensorflow as tf
import numpy as np
```

```python
# 퍼셉트론 구현
class Perceptron:
    def __init__(self,w,b):
        self.w=tf.Variable(w, dtype=tf.float32) #가중치를 입력받아 저장하는 코드
        self.b=tf.Variable(b, dtype=tf.float32) #편향 또한 저장
        
    def __call__(self, x):
        return tf.sign(tf.reduce_sum(self.w*x)+self.b) #퍼셉트론의 출력은 기본적으로 입력값에 가중치를 곱한 값을 출력함
```

- sign 함수를 씌우는 이유 : -1또는 1의 값을 가지기 위함

```python
# 유틸리티 함수 구현. 코드 꿀팁임
def v(*args):
    return np.array(args) #자동 벡터화 함수
```

```python
# 퍼셉트론 weight, bias 정의
w=v(1,1)
b=0.5

perceptron=Perceptron(w,b)
```

```python
# 퍼셉트론 동작 확인
p1=perceptron(v(1,1)) #1사분면. T T
p2=perceptron(v(-1,1)) #2사분면. F T
p3=perceptron(v(-1,-1)) #3사분면. F F
p4=perceptron(v(1,-1)) #4사분면. T F

print(p2.numpy(), p1.numpy())
print(p3.numpy(), p4.numpy())
```

- or 함수가 정상적으로 만들어졌음을 알 수 있다.

## 실습 2 : XOR 함수 구현하기

```python
# XOR 함수 구현
# 미션 1 : nand 함수 구현하기
p_nand=Perceptron(w=v(-1,-1),b=0.5)
# 미션 2 : or 함수 구현하기
p_or=Perceptron(w=v(1,1),b=0.5)
# 미션 3 : 이 둘을 이을 and 함수 구현하기
p_and=Perceptron(w=v(1,1),b=-0.5)

def xor(x):
    h1=p_nand(x)
    h2=p_or(x)
    return p_and(v(h1,h2))
```

```python
p1=xor(v(1,1)) #1사분면. T T
p2=xor(v(-1,1)) #2사분면. F T
p3=xor(v(-1,-1)) #3사분면. F F
p4=xor(v(1,-1)) #4사분면. T F

print(p2.numpy(), p1.numpy())
print(p3.numpy(), p4.numpy())
```

- xor 함수 또한 정상적으로 출력됨을 알 수 있다.
