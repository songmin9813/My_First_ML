### 회귀분석이란
- 지도 학습(supervised learning)
  - $Y=f(x)$에 대하여 입력 변수 $(X)$와 출력 변수$(Y)$의 관계에 대하여 모델링하는 것
  - Y에 대하여 예측 또는 분류하는 문제
    - 회귀(regression) : 입력 변수 $X$에 대해서 연속형 출력 변수 $Y$를 예측
    - 분류(classification) : 입력 변수 $X$에 대해서 이산형 출력 변수 $Y$(class) 예측
  - $Y=f(X)$에서 $Y$가 연속형 변수라면 $X$는 연속/이산형 변수, $f()$는 회귀 모형
  - $Y=f(X)$에서 $Y$가 이산형 변수(클래스)라면 $X$는 연속/이산형 변수, $f()$는 분류 모형

- 회귀분석
  - 입력 변수인 X의 정보를 활용하여 출력 변수인 Y를 예측하는 방법
  - 회귀분석 중 간단한 방법으로는 선형회귀분석(좌)이 있고, 이를 바탕으로 더 복잡한 회귀분석(우)가 개발

  - 대부분의 분류모델(SVM, Decision Tree)에서도 회귀분석 사용이 가능

- 단순 선형 회귀분석
  - $Y=\beta_0+\beta_1X+\epsilon$
  - 입력 변수가 X, 출력 변수가 Y일 때 단순선형회귀의 회귀식은 검은 선으로 나타낼 수 있음
  - $\beta_0$은 절편(intercept), $\beta_1$은 기울기(slope)이며 합쳐서 회귀계수(coefficients)로도 불림
  - 가정 : $\epsilon_i.~i.d N(0,\sigma^2),Y_i~N(\beta_0+\beta_1X_i,\sigma^2)$, X와 Y는 선형관계

- 단순 선형 회귀분석
- 우리가 알고 싶은 식
  - $Y=\beta_0+\beta_1X+\epsilon$
- 우리가 추정해야하는 식
  - $\hat{Y}=\hat{\beta}_0+\hat{\beta}_1X$
  - 엡실론이 빠짐
- **직선과 데이터의 차이가 평균적으로 가장 작아지는 직선**

- 추정? 실제 값과 우리가 추정한 값의 차이가 적으면 적을수록 좋을 것($e_i=y_i-\hat{y}_i$)
- 실제 값과 우리가 추정한 값의 차이를 잔차(residual)라고 하며 이를 최소화 하는 방향으로 추정

- 잔차의 제곱합(SSE; Error Sum of Squares)는 아래와 같이 표현 가능
- $SSE=\sum_{i=1}^{n}e_i^2=e_1^2+e_2^2+...+e_n^2$

- 잔차의 제곱합을 최소화시키는 이유
- 잔차의 합이 0이 되는 해는 무수히 많음(양수가 될 수 있고, 음수가 될 수도 있다.)
- 그렇다면 잔차에 절대값을 씌우니 미분이 불가능한 형태가 나옴
- 따라서 잔차의 제곱합은 미분이 가능한 형태로 유일한 해를 찾을 수 있음

- $SSE \hat{\beta}_0$과 $\hat{\beta}_0$로 편미분하여 연립방정식을 푸는 방법(Least Square Method)
- $SSE=\sum_{i=1}^{n}e_i^2=\sum_{i=1}^{n}(y_i-\beta_0-\beta_1x_i)^2$

### 회귀 계수의 해석
- $\hat{\beta}_0$의 해석- X1이 1단위 증가할 때마다 y가 $\hat{\beta}_1$만큼 증가한다.
- ex) radio 광고 예산과 매출 간의 관계
- Radio 광고 예산이 1증가할 때마다 매출은 0.2단위만큼 증가한다. 그때의 유의성은 매우 높다.
- Radio 광고 예산이 35단위일 때 예상 매출액은 9.312+0.203\*35=16.42 단위이다.

#### 선형 회귀의 정확도 평가
- 선형회귀는 잔차의 제곱함(SSE)를 최소화하는 방법으로 회귀 계수를 측정
- 즉, SSE가 작으면 작을수록 좋은 모델이라 할 수 있음
- MSE(Mean Squared Error)는 SSE를 표준화한 개념
- $RSE=MSE=\sqrt{\frac{1}{n-2}SSE}$

- $SST=SSE+SSR$의 증명
- $\sum(y_i-\bar{y})^2=\sum(y_i-\hat{y})^2+\sum(\hat{y}_i-\bar{y})^2+2\sum(\sum e_i)(\sum\hat{y}_i\bar{y})$
- $\sum(y_i-\hat{y})^2+\sum(\hat{y}_i-\bar{y})^2$
- $\sum(\sum e_i)=0$이기 때문에 뒤의 2ab부분이 날아감

- Sum of Squeares : 제곱합
- Degree of freedom : 자유도
  - Total : 분산의 자유도는 N-1임에 유의
  - Error : Regression과 Error의 합이 Total이 됨
- Mean Square : Sum of Squares를 자유도로 나눈 것

- Y의 총 변동은 회귀직선으로 설명 불가능한 변동(SSE)과 회귀직선으로 설명 가능한 변동(SSR)으로 이루어져 있음
- $R^2$는 RSE의 단점을 보완한 평가지표로 0~1의 범위를 가짐
  - 0 미만이 될수가 없음
- $R^2$은 설명력으로 입력 변수인 $X$로 설명할 수 있는 $Y$의 변동을 의미
- $R^2$이 1에 가까울수록 선형회귀 모형의 설명력이 높다는 것을 뜻함
- $R^2=\frac{SST-SSE}{SST}=1-\frac{SSE}{SST}=\frac{SSR}{SST}  where  SST=\sum_{i=1}^{n}(t_i-\bar{y})^2$
- **쓸모없는 변수가 있다면 $R^2$는 점점 증가한다는 단점**이 존재한다. 이를 염두에 두어야 함

- 회귀 분석은 **결국 Y의 변동성을 얼마나 독립변수가 잘 설명하느냐**가 중요
- 변수가 여러개일 때 각각 Y를 설명하는 변동성이 크면 좋은 변수임
  - p-value는 자연스럽게 낮아짐

### 회귀계수에 대한 검정
- 단순 선형 회귀분석의 검정
- $\hat{\beta}_1$의 표준오차 : $S.E(\hat{B_1})=\frac{\sigma}{\sqrt{S_{xx}}}$
  - 오차의 표준편차가 알려져 있지 않은 경우, $s=\sqrt{\frac{SSE}{n-2}}$를 대입하여 추정
- $\hat{\beta}_1$의 표본분포 : $t=\frac{(\hat{B_1}-B_1)}{\frac{s}{\sqrt{S_{xx}}}}$
  - 기본적으로 t분포를 사용한다.

- $\hat{\beta}_1$의 검정
  - 귀무가설: $B_1=0$(회귀계수는 0이다. 즉 변수의 설명력이 없다.)
  - 대립가설: $B_1\neq0$(회귀계수는 0이 아니다. 즉 변수의 설명력이 존재한다.)
    - 귀무가설은 기본적으로 기각하기 쉬운 것으로 설정하는 편임

- p-value가 높아지는 조건
  - 회귀계수가 낮고, 표준오차가 높아야 함
- p-value는 사실상 **t-통계량이 3이상만 넘어가도 0에 수렴**함
  - t-통계량 : 회귀계수를 표준오차로 나눈 것

- $\hat{\beta}_0$은 사실상 안 봐도 됨.
- $\hat{\beta}_1$의 신뢰 구간 (플러스 마이너스) 사이에 0이 껴져 있다면 대부분 귀무가설이 채택되는 편임

### crim(범죄율)/rm(주택당 땅수)/lstat(인구 중 하위 계층 비율) 세 개의 변수로 각각 단순 선형 회귀 분석하기

```python
#변수 설정 target/crim/rm/lsstat
target=boston[['Target']]
crim=boston[['CRIM']]
rm=boston[['RM']]
lstat=boston[['LSTAT']]
```

### target~crim 선형 회귀 분석

```python
#crim 변수에 상수항추가
crim1=sm.add_constant(crim,has_constant='add')
crim1
```

```python
#sm.OLS 적합시키기
model1=sm.OLS(target,crim1)
fitted_model1=model1.fit()
```

```python
#summary 함수 출력
fitted_model1.summary()
```

- R-squared ; y의 총 변동성에서 x(crim)이 설명하는 y의 변동성은 15퍼센트 정도임
  - 수치가 1에 가까울수록 관련성이 높다는 뜻
- 범죄율의 회귀계수값 : -0.4152, p-value값이 정말 작기 때문에 유의미함

```python
#회귀 계수 출력
fitted_model1.params
```

### Y_hat=beta0+beta1*X 계산해보기

```python
#회귀계수 * 데이터(X)
np.dot(crim1,fitted_model1.params)
```

```python
#predict 함수를 통해 yhat 구하기
pred1=fitted_model1.predict(crim1)
```

```python
#yhat과 predict 함수를 통해 구한 yhat 차이
np.dot(crim1,fitted_model1.params)-pred1
```

### 적합시킨 직선 시각화

```python
import matplotlib.pyplot as plt
plt.yticks(fontname='Arial')
plt.scatter(crim,target,label='data')
plt.plot(crim,pred1,label='result')
plt.legend()#result, data 범례(미니박스) 표시
plt.show()
```

```python
plt.scatter(target,pred1)
plt.xlabel('real_value')
plt.ylabel('pred_value')
plt.show()
```

```python
#residual 시각화
fitted_model1.resid.plot()
plt.xlabel('residual_number')
plt.show()
#균일하게 보기는 좀 어렵다
```

```python
#잔차의 합 계산
np.sum(fitted_model1.resid)
```

### rm 변수와 lstat 변수로 회귀분석 해보기

```python
#상수항 추가
rm1=sm.add_constant(rm,has_constant='add')
lstat1=sm.add_constant(lstat,has_constant='add')
```

```python
#회귀모델 적합
model2=sm.OLS(target,rm1)
fitted_model2=model2.fit()
model3=sm.OLS(target,lstat1)
fitted_model3=model3.fit()
```

```python
#모델 결과 출력
fitted_model2.summary()
```

```python
fitted_model3.summary()
```

- 아마 위의 셋 중에 하나의 지표를 고르자면 R-squared 값이 그나마 높은 lstat을 고르지 않을까 싶음

```python
#predict 함수를 통해 yhat 구하기
pred2=fitted_model2.predict(rm1)
pred3=fitted_model3.predict(lstat1)
```

```python
#시각화
plt.yticks(fontname='Arial')
plt.scatter(rm,target,label='data')
plt.plot(rm,pred2,label='result')
plt.legend()#result, data 범례(미니박스) 표시
plt.show()
```

```python
plt.scatter(crim,pred2)
plt.xlabel('real_value')
plt.ylabel('pred_value')
plt.show()
```

```python
#residual 시각화
fitted_model2.resid.plot()
plt.xlabel('residual_number')
plt.show()
```

```python
plt.yticks(fontname='Arial')
plt.scatter(lstat,target,label='data')
plt.plot(lstat,pred3,label='result')
plt.legend()#result, data 범례(미니박스) 표시
plt.show()
```

```python
plt.scatter(lstat,pred2)
plt.xlabel('real_value')
plt.ylabel('pred_value')
plt.show()
```
