### 통계적 추론
- 점추정(Point estimation)
  - 추정량을 통해 모수를 추정
  - ex) $\bar{X},s^2 \rightarrow \mu,\sigma^2$
  - 추정량의 중심점이 얼마나 근접한지는 알 수가 없다.
- **구간 추정**
  - 일정 신뢰수준 하에서 모수를 포함할 것으로 예상되는 구간을 제시
  - 신뢰 수준과 구간의 길이는 반비례

- 자료 하나당 오른쪽 그림과 같은 구간을 추정할 수 있다.
- 이때, 자료가 많아질수록 구간 추정이 그렇게 많아질 것.
- 신뢰수준이 95% : 100번의 자료분석 중 5개의 구간 추정이 $\mu$값과 일치하지 않는다는 것을 의미

- **대립가설(H1)**
  - 입증하여 주장하고자하는 가설
- **귀무가설(H0)**
  - 대립가설의 반대가설
  - 귀무가설이 아니라는 충분한 증거를 데이터로부터 보임으로써 대립가설을 입증
  - 귀무가설 하에서 통계량의 분포를 아는 것이 검정의 핵심
  

- ex) 우리반 전체 키의 평균이 160cm라는 가설 -> **귀무가설** $H_0:\mu=160$
- 우리는 이를 입증하기 위해 **대립가설**을 세울 수 있다 -> $H_1:\mu \neq 160$
- 귀무가설을 기준으로 분포를 세우고, **우리가 만든 통계량 값이 귀무가설가 얼마나 떨어져있는지에 따라 기각/채택 여부 판단**

### 오류의 종류
- 1종 오류
  - 귀무가설이 맞을 때, 귀무가설을 기각하는 오류
  - 이 오류를 최대한 범하지 않기 위해 노력함
- 2종 오류
  - 귀무가설이 틀렸을 때, 귀무가설을 기각하지 않는 오류

- **검정 통계량**
  - 표본에서 구해낼 수 있는 함수. 이 값을 기준으로 귀무가설 기각여부를 결정
- **기각역**
  - 검정통계량이 취하는 구간 중 귀무가설을 기각하는 구간
  - 적당하게(ㅅㅂ?) 잡아야 됨. 기각역을 너무 넓게(임계값을 귀무가설 중심점과 가깝게) 잡아버리면 1종 오류를 범할 확률이 높아진다.
    - 뭐만하면 기각하니까 많은 대립가설이 필요하지 않겠어?
  - 그 반대상황이라면 2종 오류를 범할 확률이 반대로 높아짐
  - 알파값이 정해지면 같이 정해지는 구간
    - 이 영역에 검정통계량이 들어가있으면 이 귀무가설을 기각할 것이다 판단 

- **알파값(유의확률)을 미리 정해주면 기각역이 결정이 된다. 단측/양측 검정에 따라서도 기각역이 정해진다.**
  - 단측검정 : 자신의 목적에 따라 한쪽 방향에 따라서만 기각을 하는 방식(몇 이상 이상 크다면 기각)
  - 양측검정 : 160에서 얼마나 차이가 나는지 판단.(양의 방향/음의 방향의 방향에 대해서는 상관이 없음) 

- **유의확률(P-value)**
  - 주어진 검정통계량을 기준으로 해당 값보다 대립가설을 더 선호하는 검정통계량이 나올 확률
  - 이 값이 유의수준(알파값)보다 낮으면 귀무가설을 기각.
  - P-value(유의확률,알파값)은 일반적으로 0.05를 세팅하는 편

**1. 기각역보다 검정통계량이 오른쪽에 있으면 귀무가설을 기각**



**2. 우리가 설정해놓은 P-Value(유의수준/알파값)보다 낮다면 귀무가설을 기각**
- **2번의 경우를 더 많이 사용하게 될 것**

### 검정통계량과 관련된 분포
- **Z 통계량**
  - 귀무가설 : X의 평균이 $\mu_0$이다.
  - $Z=\frac{\bar{X}-\mu_0}{\sqrt{\frac{\sigma^2}{n}}} \sim N(0,1)$ : 표본 정규 분포로 변환할 수 있다는 이야기
    - 이 때, 관측치의 수가 충분하다면(30개 이상, 자유도가 1) $\sigma^2$를 $s^2$로 대체 가능(모분산->표본분산)
  - 충분한 관측치를 사용하지 못했을 경우 아래의 t분포를 사용
  
  
- **t분포**
  - $t=\frac{\bar{X}-\mu_0}{\sqrt{\frac{s^2}{n}}} \sim t(n-1)$
  - 자유도가 커질수록 정규분포에 근사

- **카이제곱 분포**
  - $Z \sim N(0,1)$일 때(Z 통계량이 표준정규분포를 따른다고 가정)
    - $Z^2 \sim X^2_{(1)}, \sum_{i=1}^{k}Z^2_{(i)} \sim X^2_{(k)}$
    - Z 통계량을 제곱하게 되면 자유도가 1인 카이제곱 분포라는 소리
    - 자유도가1인 카이제곱 분포를 k개 더하면 자유도가 k인 카이제곱 분포라는 소리
  - $f(x;k)=\frac{1}{s^{k/2}\gamma(k/2)}x^{k/2-1}e^{-x/2}, x \geq 0$
    - 알 필요 없다. 외울 필요 없다. 그냥 이런게 있다.

- **전체적으로 카이제곱 분포의 자유도가 높아질수록 왜도가 낮아지는 것을 확인할 수 있다.**

- **F 분포**
  - 두 확률변수 $V_1,V_2$가 자유도 $k_1,k_2$이고 서로 독립인 카이제곱 분포를 따를 때,
    - $F=\frac{V_1/k_1}{V_2/k_2} \sim F(k_1,k_2)$
  - **확률변수의 제곱합을 관측치로 나눈 것의 비율로 이루어진 통계량**
  - 두 카이제곱 분포의 자유도가 크고, 둘의 값의 차이가 없을 수록 둘의 왜도가 낮아지는 특징을 가진다.(아래의 초록 그래프 참고)
