## 시작 전 기본 세팅

```python
import os
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
boston=pd.read_csv("C:/Users/mitha/OneDrive/바탕 화면/Boston_house.csv")
boston.head()
```

```python
# boston data에서 crim,rm, lstat 변수만 뽑아오기
x_data=boston[['CRIM','RM','LSTAT']]
x_data.head()
```

```python
#target 제외한 데이터만 뽑기
boston_data=boston.drop(['Target'],axis=1)
```

```python
#변수 설정 target/crim/rm/lsstat
target=boston[['Target']]
crim=boston[['CRIM']]
rm=boston[['RM']]
lstat=boston[['LSTAT']]
```

```python
#상수항 추가
x_data1=sm.add_constant(x_data,has_constant='add')
```

```python
#회귀모델 적합
multi_model=sm.OLS(target,x_data1)
fitted_multi_model=multi_model.fit()
```

```python
#결과출력
fitted_multi_model.summary()
```

## crim, rm, lstat, b, tax, age, zn, nox, indus 변수를 통한 다중선형회귀분석

```python
#boston data에서 원하는 변수만 뽑아오기
x_data2=boston[['CRIM','RM','LSTAT','B','TAX','AGE','ZN','NOX','INDUS']]
x_data2.head()
```

```python
#상수항 추가
x_data2_=sm.add_constant(x_data2,has_constant='add')
```

```python
#회귀모델 적합
multi_model2=sm.OLS(target,x_data2_)
fitted_multi_model2=multi_model2.fit()
```

```python
fitted_multi_model2.summary()
```

```python
#세 변수만 추가한 모델의 회귀 계수
fitted_multi_model.params
```

```python
#full 모델의 회귀계수
fitted_multi_model2.params
```

- **각각의 X가 Y를 설명하는 데에 중복성을 가져갈 수 없기 때문에 각 값의 절대값이 조금씩 줄어든 것을 확인할 수 있다.**
- 6개의 변수가 추가됨에 불과하고 R squared 값이 0.6 차이라면 변수의 개수가 적은 쪽을 선택할 것임
  - 아니면 full 변수에서 p-value가 높은 값들을 하나씩 없애보면서 최적의 값을 찾아보도록 노력하든가

```python
#base 모델과 full 모델의 잔차 비교
import matplotlib.pyplot as plt
fitted_multi_model.resid.plot(label='full')
fitted_multi_model2.resid.plot(label='full_add')
plt.legend()
```

- 잔차의 차이가 별로 안 나는 것을 확인

## VIF를 통한 다중공선성 확인

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

vif=pd.DataFrame()
vif["VIF Factor"]=[variance_inflation_factor(x_data2.values,i)for i in range(x_data2.shape[1])]
vif["features"]=x_data2.columns
#vif 추출
```

```python
vif
```

- 한 변수를 y로 두고 나머지로 회귀를 돌렸을 때, 나머지 변수들로 이 변수를 설명하는 비율
- 10 이상인 값이 굉장히 많아 보임

```python
#nox 변수 제거후(x_data3) VIF 확인
vif=pd.DataFrame()
x_data3=x_data2.drop('NOX',axis=1)
vif["VIF Factor"]=[variance_inflation_factor(x_data3.values,i)for i in range(x_data3.shape[1])]
vif["features"]=x_data3.columns
vif
```

- 그래도 조금 남아있는 것 같은데, RM에서 특히 눈에 띄는 vif 값이 보임

```python
#RM 변수 제거후(x_data3) VIF 확인
vif=pd.DataFrame()
x_data4=x_data3.drop('RM',axis=1)
vif["VIF Factor"]=[variance_inflation_factor(x_data4.values,i)for i in range(x_data4.shape[1])]
vif["features"]=x_data4.columns
vif
```

- 이전에 비해서는 vif 계수가 많이 떨어짐. 여기서 적합을 시키는 것이 좋다고 판단

```python
#x_data3, x_data4 상수항 추가 후 각 회귀 모델 적합
x_data3_=sm.add_constant(x_data3,has_constant='add')
x_data4_=sm.add_constant(x_data4,has_constant='add')
model_vif=sm.OLS(target,x_data3)
fitted_model_vif=model_vif.fit()
model_vif2=sm.OLS(target,x_data4)
fitted_model_vif2=model_vif2.fit()
```

```python
fitted_model_vif.summary()
```

```python
fitted_model_vif2.summary()
```

- AIC,BIC(낮을수록 좋음), 변수 제거 후 R squared 차이 등을 고려하여 판단
- **RM 제거로 인해 5퍼센트 정도 감소했다는 것은 굉장히 큰 차이임 제거 안 한 버전이 더 나았을 듯**

## 학습/검증데이터 분할

```python
from sklearn.model_selection import train_test_split
X=x_data2_ #상수항이 추가된 9개의 데이터
y=target
train_x,test_x,train_y,test_y=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
#random_state는 일종의 시드라고 생각하는 편이 편함
print(train_x.shape,test_x.shape,train_y.shape,test_y.shape)
```

```python
#train_x 회귀모델 적합
train_x.head()
fit_1=sm.OLS(train_y,train_x)
fit_1=fit_1.fit()
```

```python
#검증데이터에 대한 예측값과 true값 비교
plt.plot(np.array(fit_1.predict(test_x)),label='pred')#검증데이터에 대한 잔차
plt.plot(np.array(test_y),label='true')#학습데이터에 대한 잔차
plt.legend()
plt.show()
```

- 그래도 학습데이터가 검증데이터를 잘 따라가고 있구나를 알 수 있다.

```python
# x_data3학습 검증데이터 분할
X=x_data3_
y=target
train_x2,test_x2,train_y2,test_y2=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
#random_state는 일종의 시드라고 생각하는 편이 편함
print(train_x.shape,test_x.shape,train_y.shape,test_y.shape)
```

```python
# x_data4 학습 검증데이터 분할
X=x_data4_
y=target
train_x3,test_x3,train_y3,test_y3=train_test_split(X,y,train_size=0.7,test_size=0.3,random_state=1)
#random_state는 일종의 시드라고 생각하는 편이 편함
#시드를 같게 해주면 분리하는 방법이 동일하게 되면 동일한 인덱스를 가지게 됨
print(train_x.shape,test_x.shape,train_y.shape,test_y.shape)
```

```python
test_y.head()
```

```python
test_y2.head()
```

```python
#train_x3,4 회귀모델 적합
fit_2=sm.OLS(train_y2,train_x2)
fit_2=fit_2.fit()
fit_3=sm.OLS(train_y3,train_x3)
fit_3=fit_3.fit()
```

```python
#true 값과 예측값 비교
#검증데이터에 대한 예측값과 true값 비교
plt.plot(np.array(fit_2.predict(test_x2)),label='pred1')#검증데이터에 대한 잔차
plt.plot(np.array(fit_3.predict(test_x3)),label='pred2')#검증데이터에 대한 잔차
plt.plot(np.array(test_y2),label='true')#학습데이터에 대한 잔차
plt.legend()
plt.show()
```

```python
#full 모델 추가해서 비교
plt.plot(np.array(fit_1.predict(test_x)),label='pred')#검증데이터에 대한 잔차
plt.plot(np.array(fit_2.predict(test_x2)),label='pred_vif')#검증데이터에 대한 잔차
plt.plot(np.array(fit_3.predict(test_x3)),label='pred_vif2')#검증데이터에 대한 잔차
plt.plot(np.array(test_y2),label='true')#학습데이터에 대한 잔차
plt.legend()
plt.show()
```

- vif로 하나만 없앤게 true와의 차이가 별로 없음을 알아낼 수 있음

```python
#실제값과 예측값의 차이(잔차) 비교
plt.plot(np.array(test_y2['Target']-fit_1.predict(test_x)),label='pred_full')#검증데이터에 대한 잔차
plt.plot(np.array(test_y2['Target']-fit_2.predict(test_x2)),label='pred_vif')#검증데이터에 대한 잔차
plt.plot(np.array(test_y2['Target']-fit_3.predict(test_x3)),label='pred_vif2')#검증데이터에 대한 잔차
plt.legend()
plt.show()
```
