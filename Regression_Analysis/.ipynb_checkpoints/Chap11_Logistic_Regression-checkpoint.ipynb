## 로지스틱 회귀란
- 출력 변수를 직접 예측하는 것이 아니라, 두 개의 카테고리를 가지는 binary 형태이 출력 변수('성공','실패' 또는 '예','아니오')를 예측할 때 사용하는 회귀분석 방법
- $p=P(success|k개의 입력변수)$
- $1-p=(fail|k개의 입력변수)$
- 로지스틱회귀에서는 k개의 입력변수를 사용하여 성공 실패를 예측하기 위해 성공확률 $p(X)$를 모델링 함
  - $p(X)=P(success|X_1,...,X_k)=\beta_i+\beta_1X_1+...+\beta_kX_k+\epsilon$
- 방정식의 왼쪽의 범위는 $[0,1]$이지만, 오른쪽의 범위는 $[-\infty,+\infty]$이므로 다른 형태로 모델링해야함

- 따라서 왼쪽항에 자연 로그를 취해줌으로 $ln(p(X))$는 $[-\infty,+\infty]$가 됨.
  - $ln(p(X))=P(success|X_1,...,X_k)=\beta_i+\beta_1X_1+...+\beta_kX_k+\epsilon$
- 하지만 이를 만족하기 위해서는 $p(X)$가 $[0,+\infty]$의 범위여야 함
- 확률 $p(X)$의 최대값은 1이므로 $ln(p(X))$가 $+\infty$값을 가질 수 없음. 따라서 왼쪽 식을 아래와 같이 대체함
  - $logit=ln(\frac{p(X)}{1-p(X)})=\beta_i+\beta_1X_1+...+\beta_kX_k+\epsilon$

- X는 입력변수, Y는 출력변수가 1이 될 확률일 때 식은 아래와 같이 정리됨
  - $Y=p(X)=\frac{e^{\beta_0+\beta_1X+...+\beta_kX_k}}{1+\beta_0+\beta_1X+...+\beta_kX_k}=\frac{e^x}{1+e^x}$

- 좌 : 입력변수가 1개일 때, 우 : 입력변수가 2개일 때

### 로지스틱 회귀 예제
- 반도체 공정에서 특정 변수 값을 가지고 정상이냐 불량이냐를 분류하는 문제
- Y는 정상, 불량 2가지 범주를 가지는 출력 변수이며, 입력 변수로는 반도체 공정의 특정변수(pressure)를 사용하여 예측한 예제임
  - 좌 : 단순선형회귀를 이용했을 때, Pressure에 따른 불량확률에 음수가 나올 수 있음
  - 우 : 로지스틱 회귀를 이용했을 때 0~1사이의 범위를 가지는 것을 알 수 있음
- Pressure값이 약 2000정도일 때 불량 확률이 0.5가 되므로 2000을 기준으로 불량을 결정할 수 있음
