- 평균 변화율
  - $\frac{f(b)-f(a)}{b-a}$
- 순간 변화율 $f'(a)$
  - 평균 변화율의 극한 값
  - b점이 a점으로 한없이 가까워질 때, a점에서의 순간변화율
  - a점에서의 접선의 기울기

- **미분의 개념**
  - $f(x)=c \rightarrow f'(x)=0, c는 상수$
  - $f(x)=x^n \rightarrow f'(x)=nx^{n-1}, n은 자연수$
  - (참고)$f(x)=x^k \rightarrow f'(x)=kx^{k-1}, k는 0이 아닌 유리수$
  - $\{cf(x)\}'=cf'(x)$
  - $\{f(x) \pm g(x)\}'=f'(x) \pm g'(x)$
- **곱의 미분**
  - $\{f(x)g(x)\}'=f'(x)g(x)+f(x)g'(x)$
- **합성함수 미분 - 이거 약간 중요함**
  - $f(g(x))'=f'(g(x))g'(x)$
  - (증명) : $f(g(x))'=\frac{df(g(x))}{dx}=\frac{df(g(x))}{dg(x)} \times \frac{dg(x)}{dx}=f'(g(x)) \times g'(x)$
  - f(x) 부분이 크게 log함수나 exp 함수 꼴로 나타나는 경우가 많음에 유의

- **지수함수 미분**
  - $\{e^x\}'=e^x$
  - $\{a^x\}'=a^xlna$
- **로그함수 미분**
  - $\{lnx\}'=\frac{1}{x}$
  - $\{log_ax\}'=\frac{1}{xlna}$
  - (증명) : $log_ax=\frac{lnx}{lna} \rightarrow (\frac{lnx}{lna})'=\frac{1}{lna}(lnx)'$

- **미분의 활용**
- **극대값, 극소값**
  - 미분값은 접선의 기울도함수를 통하여 미분가능한 함수의 극대값, 극소값을 구할 수 있음
  - 해가 양이라면 $f'(x)=0$일 때, 극대값을 구할 수 있음
  - **확률상에서 likelihood와 밀접한 관련이 있음**

### Likelihood function(가능도 함수/우도함수)

- 왼쪽 : $\mu$가 주어져 있을 때, **x에 대한 함수(확률분포)**
- 오른쪽 : x가 주어져 있을 때, $\mu$**에 관한 함수(가능도함수)**
- $X_1=1,X_2=2,X_3=3$개의 자료가 있다고 하면, 평균 $\mu_0$는 어떤 값이 가능성이 가장 높을까?

- $\mu_0$를 추정하는 방법
- R.A. Fisher
  - "The disired probability distribution is the one that makes the observed data 'most likely'"
  - 가장 많이 관찰되는 자료를 모수로 두는 것이 타당하지 않겠는가?
  - $f(x_1,x_2,x_3)=(\frac{1}{\sqrt{2\pi\cdot1}})^3exp(-\frac{(x_1-\mu_0)^2+(x_2-\mu_0)^2+(x_3-\mu_0)^2}{2\cdot1})$

- **확률분포함수(Probability distribution function;PDF)**
  - 모수를 알 때, 확률변수의 실현값을 예측하고자 함
  - 종류
    - **확률밀도함수(Probability density function;PDF)** : 연속형 확률변수의 확률분포함수
      - 우리가 확률분포하면 생각나는 그거. 곡선이 아주 매끄러움(연속된 값을 가짐)
    - **확률질량함수(Probability mass function;PMF)** : 이산형 확률변수의 확률분포함수
      - $f(x)=P(X=x)$
    - **누적분포함수(Cumulative distribution function)** : 누적 확률분포함수
      - $F(x)=P(X \leq x)$
  - PDF라고 설명을 하였을 경우, 상황에 따라 알아서 구분을 하여야 함
- **가능도함수(Likelihood function)**
  - 확률변수의 실현값을 알 때(데이터가 있을 때), 모수를 추정하고자 함

- **Probability density function**
  - 평균 $\mu_0$, 분산 1을 독립정규분포로 따르는 확률변수 $X_i$의 확률분포함수(확률밀도함수)
    - $f(x_i)=\frac{1}{\sqrt{2\pi\cdot1}}exp(-\frac{(x_i-\mu_0)^2}{2\cdot1})$
  - $X_1=x_1, X_2=x_2, X_3=x_3,$ 3개의 자료가 있을 때, 확률분포함수
    - $f(x_1,x_2,x_3)=(\frac{1}{\sqrt{2\pi\cdot1}})^3exp(-\frac{(x_1-\mu_0)^2+(x_2-\mu_0)^2+(x_3-\mu_0)^2}{2\cdot1})$
    - 바로 위의 식에서 1~3 숫자만 바꿔가면서 곱해주면 됨(각 사건은 독립사건이기 때문에)

- **Likelihood function**
  - 동일한 함수이나, $\mu_0$을 변수로 인식한다.
    - $L(\mu_0;x_1,x_2,x_3)=(\frac{1}{\sqrt{2\pi\cdot1}})^3exp(-\frac{(\mu_0-x_1)^2+(\mu_0-x_2)^2+(\mu_0-x_3)^2}{2\cdot1})$
    - exp 내의 $x_i$**의 값과**$\mu_0$**의 위치가 반대로 되어있음을 알 수 있다.**(어떤 변수가 주체인지에 따라 달라지는 듯)
    - 이 $\mu$의 값을 추정하기 위해 가장 값이 많이 나오는 함수**(극대값ㄷ)**을 기준으로 $\mu$를 찾아내겠다는 취지에서 출발

### MLE(Maximum Likelihood Estimator) - 진심 개중요함
 - **Likelihood를 최대로 만드는 모수의 값**
   - $L(\mu_0;x_1,x_2,x_3)=(\frac{1}{\sqrt{2\pi\cdot1}})^3exp(-\frac{(\mu_0-x_1)^2+(\mu_0-x_2)^2+(\mu_0-x_3)^2}{2\cdot1})$
   - 주로 모수에 hat을 붙여$(\hat{u_0})$으로 표현
 - $(\hat{u_0})$가 최대값을 가지는 $\mu$ -> 극대값을 가지는 $\mu$ -> **Likelihood를 $\mu$로 미분해서 0을 만드는 값**

- **$\mu_0$ MLE 구하기**
  - **Likelihood 구성**
    - $L(\mu_0;1,2,3)=(\frac{1}{\sqrt{2\pi\cdot1}})^3exp(\frac{(\mu_0-1)^2+(\mu_0-2)^2+(\mu_3-1)^2}{2\cdot1})$
  - **미분하기 용이하도록 log Likelihood 구성 : 이유-log함수는 단조함수(x가 증가함에 따라 y도 증가함)이기 때문임**
    - $logL(\mu_0;1,2,3)=-\frac{2}{3}log(2\pi)-\frac{1}{2}((\mu_0-1)^2+(\mu_0-2)^2+(\mu_3-1)^2)$
    - (기본 증명 1) : $loga^b=bloga$ : 로그의 승수는 앞으로 뺄 수 있음
    - (기본 증명 2) : $log \cdot exp(c)=c$ : 로그와 exp는 만나면 서로 상쇄됨(역함수 관계)
  - **미분 실행**
    - $\frac{\delta logL(\mu_0;1,2,3)}{\delta\mu_0}=-\frac{1}{2}(2(\mu_0-1)+2(\mu_0-2)+2(\mu_0-3))$
      - $\mu_0$와 상관없는 $-\frac{2}{3}log(2\pi)$ 부분은 완전 날아가버림
  - **미분한 함수가 0이 되게 하는 $\hat{\mu_0}$을 구해냄**
    - $-\frac{1}{2}(2(\hat{\mu_0}-1)+2(\hat{\mu_0}-2)+2(\hat{\mu_0}-3))=0$
    - $3\hat{\mu_0}=1+2+3$
    - $\hat{\mu_0}=2$
    - **SHIIIIIIIIT**

### MLE를 배우는 방법에 대한 상기
- 미분을 배우는 이유는 MLE를 구하기 위해서임
- MLE는 우리가 전체적으로 모집단의 일부인 표본을 가지고 있음.
- 자료를 가장 잘 설명하는 모수를 추정하기 위해 Likelihood 개념을 사용한다.
- Likelihood를 가장 잘 설명하는 최대값을 MLE라 하는 것

- **$\sigma^2$ MLE 구하기**
  - **Likelihood 구성, $\lambda=\sigma^2$**
    - $L(\mu_0;1,2,3)=(\frac{1}{\sqrt{2\pi\cdot\lambda}})^3exp(\frac{(\mu_0-1)^2+(\mu_0-2)^2+(\mu_3-1)^2}{2\cdot\lambda})$
  - **미분하기 용이하도록 log Likelihood 구성 : 이유-log함수는 단조함수(x가 증가함에 따라 y도 증가함)이기 때문임**
    - $logL(\lambda;1,2,3,\mu_0)=-\frac{2}{3}log(2\pi\lambda)-\frac{1}{2\cdot\lambda}((\mu_0-1)^2+(\mu_0-2)^2+(\mu_0-3)^2)$
    - (기본 증명 1) : $loga^b=bloga$ : 로그의 승수는 앞으로 뺄 수 있음
    - (기본 증명 2) : $log \cdot exp(c)=c$ : 로그와 exp는 만나면 서로 상쇄됨(역함수 관계)
  - **미분 실행**
    - $\frac{\delta logL(\lambda;1,2,3,\mu_0)}{\delta\lambda}=-\frac{3}{2}\cdot\frac{1}{\lambda}+\frac{1}{2}(\lambda)^{-2}((\mu_0-1)^2+(\mu_0-2)^2+(\mu_0-3)^2)$
  - **미분한 함수가 0이 되게 하는 $\hat{\lambda}$를 구해냄**
    - $-\frac{3}{2}\cdot\frac{1}{\lambda}+\frac{1}{2}(\lambda)^{-2}((\mu_0-1)^2+(\mu_0-2)^2+(\mu_0-3)^2)=0$
    - $\hat{\lambda}=\frac{1}{3}((\mu_0-1)^2+(\mu_0-2)^2+(\mu_0-3)^2)$
