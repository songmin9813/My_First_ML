## 차원의 저주
- 각 변수의 50%영역에 해당하는 자료를 가지고 있다 할 때, 전체 자료의 얼마만큼을 확보할 수 있는가?

- **차원이 많아지게 되면 축적 데이터 비율 확보되는 자료의 양이 적어지게 된다.**
- 관측치의 수는 한정되어 있음
  - 차원이 커질수록 한정된 자료는 커진 차원의 패턴을 잘 설명하지 못한다.
  - 차원이 증가함에 따라 model complexity가 기하급수적으로 높아짐

- 가까이에 있는 변수가 가지는 값을 예측값으로 하는 모델이 있다 하자(k-Nearest Neighborhood)
  - 쓸데없는 변수가 추가되는 것은 모델의 성능에 매우 악영향을 끼침
    - 상관계수가 추가되는 것은 모델의 성능에 매우 악영향을 끼침
      - 상관계수가 매우 큰 서로 다른 독립변수, 예측하고자 하는 변수와 관련이 없는 변수

## 차원축소법
- 상관계수가 높은 변수 중 일부를 분석에서 제외?
  - 정보의 손실 발생. ex) 상관계수가 0.8이라고 하면, 0.2에 해당하는 정보는 버려지게 됨
  
  
- 차원을 줄이면서 정보의 손실을 최소화하는 방법
  - Principal component 활용
  
  
- 이외의 방법
  - 변수 선택법 : 어느 변수를 사용할지 
  - penalty 기반 regression : 전체 자료가 지나다니는 길을 제한
  - convolutional neural network : 커널을 통해 전체 차원을 조금씩 줄여나감
  - drop out & bagging : 간단한 모델 여러개를 만들면서 그 안 변수는 조금밖에 안 들어있을 때

## 공분산 행렬
- 공분산 행렬(Covariance matrix)의 정의
  - X_1,X_2이 음의 상관관계를 가지므로, 둘의 공분산은 음수일 것. X가 centering 되어 있다면

  - X가 centering되어 있다면, $Cov(X)=(X^TX)/(n-1)$

- 공분산의 형태 파악
  - 점과 내적연산을 할 경우, **점의 위치를 이동시켜 해당 공분산 구조와 비슷한 형태**를 가질 수 있다.
  - 점을 늘릴 수 있다는 소리

- 특정 방향으로 늘리는 효과를 줄 수 있음 -> **기존 값에서 가지는 타원형과 굉장히 비슷한 형태를 띄게 됨**

### 몇몇 특이 케이스 - 기억 안 해도 됨 
  - 대칭행렬이지만, 일반적인 공분산이 아닌 경우
  
  1. Positive definite가 아님

  2. 행렬식이 0인 경우(AD-BC=0)

## Principle Components의 개념
- 차원을 줄이면서 정보의 손실을 최소화 하는 방법
  - 더 적은 개수로 데이터를 충분히 잘 설명할 수 있는 새로운 축을 찾아냄

- **PC를 얻어낸다는 것**
  - 공분산이 데이터의 형태를 변형시키는 방향의 축과 그것에 직교하는 축을 찾아내는 과정
  - 2차원의 경우 공분산이 나타내는 타원의 장축과 단축

- Principle components = PC = PC score
  - 찾아낸 새로운 축에서의 좌표값을 의미
  - 새로운 축에 내린 정사영
