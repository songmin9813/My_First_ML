## 회귀모델의 성능지표
- $R^2$
- 변수가 증가하면 증가할수록 $R^2$는 자연스럽게 증가
  - $Y$의 변동성 안의 $X$가 조금이라도 추가된다면 $R^2$는 **자연스럽게 증가할 수밖에 없음**

- $Adjusted R^2$
  - 변수 수가 증가하면 자연스게 $SSR$이 증가하고 $R2$또한 자연스레 증가함
  - 회귀분석의 성능지표로써 $R2$가 큰 의미가 없음
  - $R2$에 변수 수 만큼 penalty를 주는 지표가 $Adjusted R2$ 방식임
  - $R^2=1-\frac{SSE}{SST}$
  - $Adjusted R^2=1-\frac{\frac{SSE}{(n-p)}}{\frac{SST}{(n-1)}}$

- **AIC(Akakie information criterion)**
  - 모델의 성능지표로써 MSE에 변수 수 만큼 penalty를 주는 지표
  - 일반적으로 **회귀분석에서 Model Selection할 때 많이 쓰이는 지표**
  - $AIC=nln(\frac{SSE}{n})+2(p+1)$

- **BIC(Bayes Information Criteria)**
  - AIC의 단점은 표본 n이 커질 때 부정확해짐
  - 이를 보완한 지표가 BIC(AIC와 그렇게 큰 차이는 없음)
  - $BIC=nln(\frac{SSE}{n})+(p+1)ln n$

### 모형의 성능지표
- **MSE(Mean Squared Error) - 가장 많이 사용하는 성능지표**
  - $f$가 제대로 추정되었는지 평가하기 위해, 예측한 값이 실제 값과 유사한지 평가하는 척도가 필요함
    - $MSE=\frac{1}{n}\sum_{i=1}^{n}[y_i-\hat{f}(x_i)]^2$
    - $y_i$ : 실제 값
    - $\hat{f}(x_i)$ : 예측값
    - $n$ : 데이터개수
    - $MSE$ : $n$개 데이터에 대한 평균 오류자승
  - $MSE$는 실제 종속 변수와 에측한 종속 변수간의 차이
  - $MSE$가 작을수록 좋지만, $MSE$를 과도하게 줄이면 과적합의 오류를 범할 가능성이 있음
  - 따라서, 검증 집합의 $MSE$를 줄이는 방향으로 $f$를 추정

- **MAPE(mean abvsolute percentage error)**
  - $MAPE = \frac{100}{n}\sum_{i=1}^{n}|\frac{y_i-\hat{x_i}}{y_i}|, (y_i \neq 0)$
    - $n$ : 데이터 개수
    - $y_i-\hat{f}(x_i)$ : 예측한 종속변수
    - $y_i$ : 실제 종속변수
  - MAPE는 퍼센트값을 가지며 0에 가까울수록 회귀 모형의 성능이 좋다고 할 수 있음
  - 0~100% 사이의 값을 가져 이해하기 쉬우므로 성능 비교 해석이 가능하다.

- **정확도(Accuracy) : Classicification에서 사용**
  - 정확도는 전체 데이터 중에서 모형으로 판단한 값이 실제값과 부합하는 비율
  - 분모는 전체 데이터가 되고 분자는 모형이 실제 정상을 정상으로 그리고 실제 이상을 이상으로 옳게 분류한 데이터임
  - $Accuracy = \frac{옯게 분류된 데이터 수}{전체 데이터의 수}$

- **정밀도, 재현율, 특이도**
  - 분류 모형의 목적에 따라 다양한 지표를 볼 수 있음
    - 정밀도 : 분류 모형이 불량을 진단하기 위해 얼마나 잘 작동했는지 보여주는 지표
    - 재현율 : 불량 데이터 중 실제로 불량이라도 진단한 제품의 비율(진단 확률)
    - 특이도 : 분류 모형이 정상을 진단하기 위해 잘 작동하는지를 보여주는 지표

- **G-meam, F1 measure**
  - 실제 데이터의 대표적인 특성에는 불량(이상) 데이터를 탐지하는 것이 중요하다는 점과 이러한 불량 데이터는 매우 소수의 데이터라는 점(class imbalanced 문제)
  - 데이터 1000개 중 불량 데이터가 10개 나머지 990개는 정상 데이터라 가정했을 때, 분류 모형이 모든 데이터를 정상 데이터가 예측해도 정확도는 99%이며(accuracy paradox), 만약 우연히 1개만 불량이라고 예측했는데, 실제 불량일 경우 정밀도 지표는 1이다.
  - 실제 데이터의 특성상 정확도보다는 제 1종 오류와 제 2종 오류 중 성능이 나쁜 쪽에 더 가중치를 주는 G-mean 지표나 불량에 관여하는 지표인 정밀도와 재현율만 고려하는 $F_1 measure$가 더 고려해볼 수 있는 지표임
  - 실제로는 **$F_1 measure$를 그나마 더 많이 쓴다고 함**

- **ROC curve, AUC**
  - 가로축을 1-특이도(specificity), 세로축을 재현율(recall)로 하여 시각화한 그래프를 ROC(Receiver Operating Characteristic) curve라 함
  - 이 때 ROC curve의 면적을 AUC라고 함 : 1에 가까울수록 좋은 성능지표임
  - 모델이 아예 가운데에 붙어있으면 쓸모없는 모형이라 판단함
