## PC score의 기하학적 해석
- $XV=UD,x_i^T=[x_{i1},...,x_{ip}],X=\begin{bmatrix}x_1^T\\\vdots\\x_n^T\end{bmatrix},V=[v_1,...,v_p]$
- i번째 관측치의 PC score=$x_i^TV=[x_i^Tv_1,...,v_i^Tv_p]$
- i번째 관측치의 1번째 PC score=$x_i^Tv_1$
  - 관측치 $a=x^T$와 eigen vector b=$v_1$의 **내적값**

- PCA 활용 - 각 PC의 의미 파악

- PC1은 아래로 갈수록 글자가 진해지고, 오른쪽으로 갈수록 꼬리 부분이 길어지는 것을 확인할 수 있었다.

## Kernel PCA
- 다음과 같은 데이터는 어떻게 처리해야 할까?
  - X1, X2 사이의 공분산은 0일 것임. 비선형적 관계

- 관측치 사이의 패턴은 존재하지만, 변수간의 선형관계가 아닐 때
  - 관측치 사이의 패턴을 수치화하고, 이것의 PC를 구해냄
    - K(Kernel matrix)는 관측치 사이의 유사도 개념(점 사이의 거리를 기준)
      - **비슷한 관측치일수록 큰 값**
      - **서로 이질적인 관측치일수록 작은 값**
      - $n \times n$matrix
- K : 두 관측치가 가까울수록 값이 커지고, 멀리 떨어져있을수록 작은 값을 가지는 **유사도 정도를 의미**
- K의 예시
  - X가 centering 되었을 때
  - $K=XX^T$또는 $K(x,x^T)=exp(-[x-x']^T[x-x'])$

- ex: $D=XX^T$인 경우
  - $K=XX^T$
  - $K=UDV^TVD^TD^T=UD^2U^T$
  - $K$의 Eigen Vectors = $U$의 칼럼벡터
  - $K$의 Eigen Values = $D^2$의 대각성분
  - PC score : $UD$로 구함
    - $X=UDV^T$
    - $XV=UD$
