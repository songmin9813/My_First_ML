## 다중공선성
- 독립변수들이 강한 선형관계에 있을 때 다중공선성이 있다고 한다.(ex: 지금까지의 newspaper 자료)

- 겹치는 설명에 대해서는 각 변수가 중복으로 가져갈 수 없음
- **따라서 매출을 설명하는 변동성이 겹침. 잘못된 변수해석, 예측 정확도 하락 등을 야기한다.**

- '이 방법이면 다중공선성을 해결할 수 있다.'같은 방법은 현재 존재하지 않는다.
- 다중공선성을 진단하는 방법, VIF(Variance inflation factor), 변수들간의 Correlation 등으로 진단
  - $VIF_i=\frac{1}{1-R_i^2}$

- 다중공선성을 해결하는 방법
- **Feature Selection : 중요 변수만 선택하는 방법**
  - 단순히 변수를 제거하는 방법(correlation 등의 지표를 보고)
  - Lasso
  - Stepwise
  - 기타 변수 선택 알고리즘(유전알고리즘 등)
- **변수를 줄이지 않고 활용하는 방법**
  - AutoEncoder 등의 Feature Extraction 기법(딥러닝기법)
  - PCA
  - Ridge
- **중요한 Feature(변수)를 뽑는 방법은 데이터사이언스 분야에서 현재까지 큰 이슈이다.**

### VIF(Variance inflation factor)
- $VIF_i=\frac{1}{1-R_i^2}$ : 지수가 10 이상인 경우 다중공선성이 있는 변수라 판단
1. $X1$을 종속변수, 나머지 변수를 독립변수로 하여 회귀모델$(f1)$ 적합
  - $x_1=\beta_2X_2+\beta_3X_3+...+\beta_pX_p+\epsilon$
2. $f1$의 $R^2$를 이용하여 $VIF_i$계산
  - $R_1^2$
3. $VIF_1$의 의미 : 다른 변수의 선형 결합으로 $X1$을 설명할 수 있는 정도
  - **다른 변수로도 X1를 충분히 설명할 수 있다. 빼도 상관없음**


- 모델의 성능이 좋아질 확률은 높아지겠지만, 보장은 하지 않음
- 중요한 도메인(변수)일 수도 있기 때문에 분석자의 판단에 맡기도록 한다.

- 상관행렬(correlation matrix)
  - 상관행렬 및 산점도를 보고 판단

- 최근 머신러닝 기법들은 중요 Feature만을 뽑는 알고리즘을 내장하고 있다.
  - 중요변수 100개 VS 모든변수 10000개라면 **중요변수 100개가 무조건 좋음**(예측성능,학습속도,활용측면)
- 머신러닝 기법은 기본적으로 학습데이터 내에서 예측력을 높이기 위해 **최대한 많은 변수를 활용**하려 함
  - 따라서 Overfitting이 발생할 확률이 높아짐
