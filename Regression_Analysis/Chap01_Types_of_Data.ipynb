## 통계학이란?

- 모집단(Population) : 연구의 대상이 되는 모든 개체들을 모은 집합
  - 시각적, 공간적 제약으로 인해 모집단 전체를 대상으로 한 분석은 불가능
- 표본(Sample) : 모집단의 일부분의 관측값들


- 모집단에서 표본을 추출해내고, 표본을 통해 모집단을 추정/추론하는 방식

- 모수(parameter) : 수치로 표현되는 모집단의 특성


- 통계량(statistic) : 표본의 관측값들에 의해서 결정되는 양


- 추정량(estimator) : 모수를 추정하고자 하는 목적을 지닌 통계량
  - 모수 : $\mu,\sigma^2$

## 자료의 종류
- 수치형(양적자료)
  - 연속형 (ex: 몸무게, 키)
    - 숫자와 숫자 사이에 항상 다른 수가 들어올 수 있다는 특징을 가진다.
  - 이산형 (ex: 전화 통화 수)
    - 숫자와 숫자 사이가 분절될 수 있는 특징을 가짐
  
  
- 범주형(질적자료)
  - 순위형(ex: 학점)
    - 기본적인 자료에 순위가 존재한다는 특징(상위 데이터가 존재할 수 있음)
  - 명목형(ex: 성별)

- 자료가 달라지면 우리가 하는 분석이 달라진다.


- 설명 변수 : X, 반응 변수 : Y
  - 설명 변수가 **범주형**, 반응 변수가 **범주형**이라면 **범주형 자료분석(카이스퀘어 검정)**
  - 설명 변수가 **연속형**, 반응 변수가 **범주형**이라면 **로지스틱 회귀분석**
  - 설명 변수가 **범주형**, 반응 변수가 **연속형**이라면 **분산분석**
  - 설명 변수가 **연속형**, 반응 변수가 **연속형**이라면 **회귀분석을 실시**한다.

## 범주형 자료의 요약 방법
- 범주형 자료 : 도수분포표, 막대/원형 그래프

- 그림을 일단 그려보고, 변수간의 관계를 통해 어떤 분석을 하게될 지 분석하는 것이 우선이다.

- 연속형 자료 : Box plot, 히스토그램(Histogram)

- Box Plot에 보이는 실선이 가장 많은 데이터의 분포를 가지는 범위임을 의미한다.
- 히스토그램은 어떤 변수가 전반적으로 어떻게 분포되어 있는지 눈으로 쉽게 확인할 수 있다.

## 자료의 요약 - 수치
- **모집단 개체**의 수 : N
- 중심 경향값(대표값)
- 전반적인 위치가 어디에 있는지(평행이동의 개념으로 이해)
  - 평균(mean) : $\mu=\frac{x_1+...+x_N}{N}=\frac{\sum_{i=1}^{N}x_i}{N}$
  - 중앙값(median) : 크기순으로 정렬시켜 중앙에 위치한 값
  - 최빈값(mode) : 가장 자주 나오는 값
    - 범주형 자료에서 많이 사용하게 됨

- **mean과 median이 같은 경우는 그래프가 평행일 때이다.**
- 한쪽으로 치우져져 있는 경우(꼬리가 긴 경우), **평균이 치우진 쪽으로 가게 되고, 최빈값은 그 반대로 치우지게 된다.**
  - 평균은 outlier(이상치)에 대해 값이 많이 변하는 특징을 가지기 때문에

- 모집단 개체의 수 : N
- **산포도(퍼진 정도)**
  - 분산(Variance) : $\sigma^2=\frac{\sum_{i=1}^{N}(x_i-\mu)^2}{N}$
  - 사분위수 범위(Inter quarile range)
    - 전체 관측값을 크기순으로 정렬했을 때 중앙에 위치한 50%의 관측치가 가지는 범위
    - outlier를 어느정도 예방할 수 있음
- **정규 분포**
  - 자연과학 현상을 설명할 때 가장 널리 쓰이는 분포
  - 위치는 평균에 의해, 모양은 분산에 의해 결정된다.
  - **평균값에 따라 중심축이 이동. 분산이 커질수록 그래프가 완만해진다.**

- 분포도
  - 왜도(Skewness)
    - 분포의 비대칭 정도
    - 왼쪽(오른쪽)으로 꼬리가 길게 늘어져 있다면 Left-skewed(Right-skewed)
    - Negative(Positive) skewed라고도 표현함
- 첨도(Kurtosis)
  - 분포의 꼬리 부분의 비중에 대한 측도
  - $K_s=0$ : 뾰족한 정도가 정규분포와 동일함
    - 만약 $K_s>0$라면, 정규분포보다 뾰족하다는 말

- 추정량(estimator) : 모수를 추정하기 위한 목적을 가진 통계량
  - 추정량의 종류
    - 표본평균 : $\bar{X}=\frac{\sum_{i=1}^{n}x_i}{n}$
    - 표본분산(Sample variance) : $s^2=\frac{\sum_{i=1}^{n}(x_i-\bar{X})}{n-1}$
      - 분자의 뮤값이 X로 바뀌어있음(전체의 평균을 모르기 때문에 표본평균으로 대체)
      - 분모가 n-1로 바뀌어있음(모집단 n이 아니기 때문에 일종의 보정을 가해준 셈)
