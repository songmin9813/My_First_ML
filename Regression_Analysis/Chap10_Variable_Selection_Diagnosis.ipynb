## 모델 선택
- 변수가 여러개일 때 최선의 변수 조합을 찾아내는 기법
- 변수의 수가 $p$개일 때 변수의 총 조합은 $2^p$로 변수 수가 증가함에 따라 변수 조합의 수는 기하급수적으로 증가
- 총 변수들의 조합 중 최적의 조합을 찾기 위한 차선의 방법
- optimal은 아님. optimal한 조합을 찾는 방법은 모든 경우의 수 조합을 다 해보는 것임
  1. Feedforward Selection 방법
  2. Backward Elimination 방버 
  3. Stepwise 방법

- **Feedforward Selction : 변수를 추가해가며 성능지표를 비교해가는 방법**
  - 변수를 하나씩 추가해가며 성능지표를 비교해가는 방법
  - AIC를 비교해가며 가장 낮을때의 값에서의 모델을 선택하도록 함

- 낮은 수치가 나온 시점에서 다른 변수의 조합은 생각하지 않고 stop하는 특징을 가짐

- **Backward Elimination : 변수를 하나씩 제거해가면서 성능지표를 비교해가는 방법**
  - 하나씩 지웍면서 AIC 지표가 최저를 찍을 때 그 모델을 선택

- **Stepwise**
  - 가장 유의한 변수를 추가하거나 유의하지 않는 변수를 제거해나가는 방법
  - Feedforward와 Backward의 방법을 혼용한 형태라 볼 수 있음
  - 전진선택법을 사용할 때 한 변수가 선택되면 이미 선택된 변수 중 중요하지 않은 변수가 있을 수 있음
    - 각 단계에서 이미 선택된 변수들의 중요도를 다시 검사하여 중요하지 않은 변수를 제거함
  - 일반적으로 가장 널리 쓰이는 방법
    1. 변수 입력/제거를 위한 p-value 임계치 설정
    2. Forward selection을 통한 변수 설정
    3. 선택된 변수 중 유의미한 변수를 남기고 제거, 2~3반복
    4. 변수가 추가되거나 제거할 케이스가 없는 경우 종료
  - 기본적인 시작 방법이 Feedforward 방법과 비슷하기 때문에 유사한 결과가 나올 수 있음

## 교호 작용(Interaction term)
- 변수간의 시너지 효과
  - X1과 X2는 Y에 영향을 끼치지는 않지만, X1과 X2가 결합됨으로 Y에 중요한 영향을 끼칠 수 있음
  - X1, X2, 그리고 X1과 X2의 교호작용에 대해서 회귀 모델 방정식은 다음과 같이 쓸 수 있음
  - 교호 작용은 일반적으로 도메인지식에 근거하여 추가하여야 함
    - 다중공선성, 모델링 능력에 영향을 미치기 때문에 다른 방법은 잘 안 쓰는 편임
  - $\hat{Y}=\hat{\beta}_1X_1+\hat{\beta}_2X_2+\hat{\beta}_3X_1X2$

- 명목형 변수(Dummy variable)
  - 성별, 대학, 지역 등 명목형 변수의 경우 전처리가 필요함
  - ex) 학력과 수입의 관계
  - 고등학교는 00, 대학교는 01, 대학원은 11(one-hot encoding)

## 회귀분석의 진단
- 적절한 변수를 통해 어느정도 성능지표가 잘 나오는 모델을 만들었다.
- 그렇다면 과연 이 회귀모델이 잘 만들어진 모델인 것인가에 대한 진단이 필요하다.
- 회귀분석에는 아래 잔차에 대한 세 가지 가정이 존재한다.
  - **정규성, 독립성, 등분산성**
  - $E_i=y_i-\hat{y}_i$
- 이 세가지 가정을 만족할 시 잘 만들어진 회귀모델이라 판단한다.
- 회귀분석의 가정 : $\epsilon_i\sim i. i. d N(0,\sigma^2), Y_i\sim N(\beta_0+\beta_1X_i,\sigma^2)$, X와 Y는 선형관계

- 회귀모델을 잘 만들었을 경우, 잔차는 정규분포를 따른다.
- 일반적으로 Residuals 산점도, Normal Q-Q Plot, Residual vs fitted plot으로 진단한다.

- **잔차가 가정에 위배된 경우**
  1. Y에 대하여 log 또는 root를 씌워줌
  2. 이상치 제거
  3. 다항 회귀 분석
