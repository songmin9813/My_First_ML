## Linear Discriminant Analysis(LDA)

- 다음 1,2,3을 잘 나누는 방법은?
  - 중간이 정규분포인 것 마냥 고르게 분포되어 있을 때

- 가정
  - 각 숫자 집단은 **정규분포의 확률분포**를 가진다.
  - 각 숫자 집단은 **비슷한 형태의 공분산 구조**를 가진다.

- LDA 결과 얻게되는 decision boundary의 특징
  - Boundary에 직교하는 축 : 자료들을 이 축에 정사영시킨 분포의 형태를 고려한다.
  - 각 정규분포의 분산 대비 평균의 차이를 구함

- **그럼 축을 어떻게 내려야 최대의 효율을 낼 수 있을까?**
  - **두 평균 vector의 차이 벡터를 이용**

- 하지만 위의 그림으로는 둘 그림의 차이가 그렇게 난다고 할 수 없다.(겹치는 구간이 너무 많음)
- 따라서 아래의 조건을 추가하여 이 둘을 동시에 달성하는 것이 목표
  - **평균의 차이는 최대화**
  - **두 분포의 각각의 분산은 최소화**
- 결국 **분산대비 평균의 차이**를 극대화하는 boundary를 찾고자 하는 것임

- 타원의 단축 방향을 찾아서 정사영 내리는 것이 분산이 최소화되는 방향임을 알 수 있다.

## LDA 접근법 첫번째 : 다변량 정규분포
- **정규분포**
  - $f(x_1)=\frac{1}{\sqrt{2\pi\sigma^2}}exp(\frac{-1}{2}(\frac{x_1-\mu_1}{\sigma_1})^2)$
  - 자주 활용하기만 하는것이지, 무조건 외울필요가 없듯, 이변량 정규분포 또한 느낌으로만 접근하면 되는 것
- **이변량 정규분포-정규분포를 3차원 공간으로 확장**
  - $f(x_1,x_2)=\frac{1}{\sqrt{(2\pi)^2\sigma_1^2\sigma_2^2(1-\rho^2)}}exp(\frac{-1}{2(1-\rho)^2}[(\frac{x_1-\mu_1}{\sigma_1})^2-2\rho(\frac{x_1-\mu_1}{\sigma_1})(\frac{x_2-\mu_2}{\sigma_2})+(\frac{x_2-\mu_2}{\sigma_2})^2])$
    - $\rho$ : correlation : $\frac{Cov(x_1,x_2)}{\sigma_1\sigma_2}$ 
      - 공분산이 0이라면(독립) 단순히 곱한 형태, 아니라면 위의 이변량 정규분포를 따른다.
  - 두 사건을 한꺼번에 표기할 수 있는 공식
- 3차원 그래프 그려보기

- 다변량 정규분포로 발전하기 위해서는 Matrix 차원에서 이해하는 것이 중요함
  - 이변량 정규분포

- 다변량 정규분포

- 시그마를 위의 공식에 대입하게 되면 같은 값이 나온다.

### 다변량 정규분포의 LDA에의 적용
- k번째 범주 자료의 분포함수 - 각 변에 log를 취한 형태

- 따라서, $p(y=k)=\pi_k$

- 여기서 k번째 범주가 나타날 사전확률을 고려하면
  - $logf_k(x,\mu_k)=-\frac{1}{2}log|\sum_{k}|-\frac{1}{2}(x-\mu_k)^T\sum_{k}^{-1}(x-\mu_k)+-\frac{p}{2}log2\pi+log\pi_k$

- k번째, l번째 범주에서 현재 자료가 나왔을 확률을 각각 구해보기
  - **어느쪽이 더 큰지 알아보기 위해 두 값을 빼는 것임**
  - 양수가 나온다면 k범주가 더 크고, 음수가 나온다면 l범주쪽으로
- $logf_l(x,\mu_l)=-\frac{1}{2}log|\sum_{l}|-\frac{1}{2}(x-\mu_l)^T\sum_{l}^{-1}(x-\mu_l)+-\frac{p}{2}log2\pi+log\pi_l$
- 만약 $\sum_{l}=\sum_{k}$라면(가정), 이 둘의 차이는(같은 값끼리 날아감)


- 여기서 찾아낸 식이 바로 LDA의 decision boundary이다.
  - $log\frac{\pi_k}{\pi_l}-\frac{1}{2}(\mu_k+\mu_l)^T\sum^{-1}(\mu_k-\mu_l)+x^T\sum^{-1}(\mu_k-\mu_l)$
  - 여기서 양수면 k범주, 음수면 l범주인 것
  

## LDA 모델 정의
- LDA decision boundary
  - 분산대비 평균의 차이를 극대화하는 boundary
- 가정
  - 각 숫자 집단은 **정규분포 형태의 확률 분포**를 가진다.
  - 각 숫자 집단은 **비슷한 형태의 공분산 구조**를 가진다.
- 확률분포 관점
  - Y가 K개의 범주를 가질 때, k=1,...,K
  - y=k일 때, 공분산 구조 $\sum_{k}$를 가지는 p개의 정규분포 변수의 분포
  - $f_k(x)=\frac{1}{(2\pi)^{p/2}|\sum_{k}|^{1/2}}exp(-\frac{1}{2}(x-\mu_k)^T\sum_{k}^{-1}(x-\mu_k))$

- **확률 관점에서 다시 보도록 함**
  - 아래의 값이 클수록, l 대비 k 범주에 속할 확률이 높아진다.
  
- 위의 식은 k와 관계없는 공통 공분산 구조 $\sum$을 가짐으로 인해, 다음과 같이 **x에 대한 1차식(linear)**으로 정리된다.
  

- 추가 : QDA(제곱의 형태) - 2차에 해당하는 형태가 소거가 되지 않은 상태-나중에 배우게 될 것임
  - 그렇다면 QDA의 전제조건은 l과 k가 공분산 형태가 아닌 경우가 되겠지?

- **아래와 같이 linear discriminant function을 정의**
  - $\delta_k(x)=x^T\sum^{-1}\mu_k-\frac{1}{2}\mu_k^T\sum^{-1}\mu_k+log\pi_k$
    - $\pi_k$:k라는 범주가 출현할 확률
  - 위에서 정의한 확률의 비율은 아래와 같이 나타낼 수 있음.
    - $log\frac{Pr(G=k|X=x)}{Pr(G=l|X=x)}=\delta_l(x)-\delta_l(x)$
  - 이 값을 가장 크게 하는 것이 k가 y의 추정된 그룹

- $\hat{y}=argmax\delta_k(x)_{(k=1,...,K)}$

## LDA 두번째 접근법 : 사영(projection)
- v의 u로의 사영(projection of $v$ onto $u$)은 아래와 같이 정의됨
  - 
- $proj_uv=\frac{v*u}{|u|^2}u$
- 만약 u의 길이가 1 단위벡터라면, 위 **벡터의 크기는 내적의 값과 일치**

## 목표: 분산은 최소화하면서 평균을 최대화하는 사영을 찾는 것
  - 일단 사영된 곳에서의 평균의 차이와 분산을 표현해야 함
  - 사영시킬 대상이 되는 벡터를 a라 가정
  - 사영된 자료 : $a^Tx_i$, 사영된 평균 : $m_1=a^T\mu_1,m_2=a^T\mu_2$
  - 사영된 표본분산
    
      - 뒤의 식 $a^T(x_i-\mu_1)(x_i-\mu_1)^Ta$는 각 스칼라 꼴을 제곱한 것과 같은 의미로 해석

- 이 때, 아래의 식으로도 표현이 가능


- $argmax\frac{a^TBa}{a^TWa}$는 다항식의 형태를 띤다.(2차항을 생각해)
  - 위의 값에서 최대(극대)값을 찾기 위해서는 **미분하여 0이 되는 값을 찾으면 된다.**
- 위의 식을 a에 대하여 미분하여 해를 구하게 되면 다음과 같아짐

- 마지막 값에서 a값 같은 거 보임? **eigen vector**와 연관성이 있다!

- (추가)-몫의 미분법 : $(\frac{g}{f})'=\frac{g'f-gf'}{f^2}$

## 사영과 Eigen vector의 연관성
- 정방행렬 A에 대하여 $Av=\lambda v$를 만족할 경우 v는 고유 벡터(eigen vector), $\lambda$를 고유값(eigen value)라고 한다.
- 마지막 값만을 가져오도록 함
- $W^{-1}Ba=(a^TWa\frac{a^TBa}{(a^TWa)^2})a$ : 소괄호 안에 있는 모든 값은 스칼라이다.
  - $A=W^{-1}B$
  - $\lambda=(a^TWa\frac{a^TBa}{(a^TWa)^2})$
  - $v=a$

## 사영과 확률모델과의 연관성


- Vector의 길이는 무시하고 방향만을 생각한다면
  - $a=W^{-1}(\mu_1-\mu_2)$
- 위의 식에서 $W$를 $\sum$으로 대체해보면, eigen vector $a$는 아래와 같다.
  - $a=\sum^{-1}(\mu_1-\mu_2)$
    - $log\frac{\pi_k}{\pi_l}-\frac{1}{2}(\mu_k+\mu_l)^T\sum^{-1}(\mu_k-\mu_l)+x^T\sum^{-1}(\mu_k-\mu_l)$
  - 뒤의 값하고 연결이 되지 않는가?
  - $a$는 바로 **찾아낸 decision boundary 식에서, 법선 vector의 역할**을 하고 있다.

## 정리
- 아이디어 : **분산을 최소화, 평균의 차이를 최대화하는 축에 수직인 boundary**를 찾고자 함
- 투영을 통해 찾아낸 새로운 축 $a=\sum^{-1}(\mu_1-\mu_2)$
- 확률 모델로 찾아낸 decision boundary
  - 
  - 붉은색 : 법선 벡터, 푸른색 : 지나는 점
- 전제조건 : $log\frac{\pi_k}{\pi_l}$이 무의미한 값을 가지게 된다.
- $Y=ax-ab$꼴
  - $Y=ax-ab=a(x-b)$ : **기울기가 $a$이고 $b$점을 지나는 직선**

## LDA의 심화적 이해
- LDA dicision boundary : 분산대비 평균의 차이를 극대화 하는 boundary
- 가정
  1. 각 숫자 집단은 정규분포 형태의 확률분포를 가진다.
  2. 각 숫자 집단은 비슷한 형태의 공분산 구조를 가진다.
- 장단점
  - **장점**
    - Naive Bayes와 달리 설명변수간의 공분산 구조를 설명할 수 있다.
      - Naive Bayes는 설명변수 간의 관계가 독립적이다는 가정이 있음
    - 가정이 위반되더라도 비교적 *robust(변화에 민감하지 않음)*함
  - **단점**
    - 가장 작은 그룹의 샘플 수가 설명변수의 개수보다 많아야 함
    - 정규분포 가정에 크게 벗어나는 경우 잘 설명하지 못함
    - **y 범주 사이에 공분산 구조가 다른 경우를 반영하지 못함**

- k와 관계없는 **공분산 구조가 다른 경우 이 가정을 버린 것이 QDA**
- Quadratic discriminant functions
  - $\delta_k(x)=-\frac{1}{2}log|\sum|-\frac{1}{2}(x-\mu_k)^T\sum_{k}^{-1}+log\pi_k$
  - X에 대한 Quadratic form - 2차 형태

- LDA, QDA와의 비교

- 중간 그림 : 식이 아닌 변수에 제곱을 취한 형태로 보는 것
  - 각각 다른 공분산 형태가 수식에 반영되어있지 않음

- 두번째 사진과 세번째 사진과의 차이
  - QDA의 상대적 장점 : y범주별로 공분산 구조를 다르게 할 수 있음
  - QDA의 상대적 단점 : 설명변수의 개수가 많을 경우, 추정해야 하는 모수가 많아짐
    - 샘플이 많이 필요

## Linear Discriminant Analysis 실습
## 1. Linear Discriminant Analysis

```python
import numpy as np
```

- LDA를 위한 함수 불러오기

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
X=np.array([[-1,-1],[-2,-1],[-3,-2],[1,1],[2,1],[3,2]])
y=np.array([1,1,1,2,2,2])
```

- LDA 모델 구축

```python
clf=LinearDiscriminantAnalysis()
clf.fit(X,y)
```

```python
print(clf.predict([[-0.8,-1]]))
```

## 2. Quadratic Discriminant Analysis

```python
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
```

- QDA 모델 구축

```python
clf2=QuadraticDiscriminantAnalysis()
clf2.fit(X,y)
```

```python
print(clf2.predict([[-0.8,-1]]))
```

- LDA, QDA 비교

```python
from sklearn.metrics import confusion_matrix
y_pred=clf.predict(X)
confusion_matrix(y,y_pred)
```

```python
y_pred2=clf2.predict(X)
confusion_matrix(y,y_pred2)
```

## 3. LDA, QDA의 시각적 비교

```python
from sklearn.datasets import make_moons, make_circles, make_classification
from matplotlib.colors import ListedColormap
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
h=0.2
names = ["LDA", "QDA"]
classifiers = [
    LinearDiscriminantAnalysis(),
    QuadraticDiscriminantAnalysis()]

X, y = make_classification(n_features=2, n_redundant=0, n_informative=2,
                           random_state=1, n_clusters_per_class=1)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [make_moons(noise=0.3, random_state=0),
            make_circles(noise=0.2, factor=0.5, random_state=1),
            linearly_separable
            ]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds in datasets:
    # preprocess dataset, split into training and test part
    X, y = ds
    X = StandardScaler().fit_transform(X)
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.4)

    x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
    y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(['#FF0000', '#0000FF'])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
    # and testing points
    ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6)
    ax.set_xlim(xx.min(), xx.max())
    ax.set_ylim(yy.min(), yy.max())
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)

        # Plot the decision boundary. For that, we will assign a color to each
        # point in the mesh [x_min, m_max]x[y_min, y_max].
        if hasattr(clf, "decision_function"):
            Z = clf.decision_function(np.c_[xx.ravel(), yy.ravel()])
        else:
            Z = clf.predict_proba(np.c_[xx.ravel(), yy.ravel()])[:, 1]

        # Put the result into a color plot
        Z = Z.reshape(xx.shape)
        ax.contourf(xx, yy, Z, cmap=cm, alpha=.8)

        # Plot also the training points
        ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright)
        # and testing points
        ax.scatter(X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright,
                   alpha=0.6)

        ax.set_xlim(xx.min(), xx.max())
        ax.set_ylim(yy.min(), yy.max())
        ax.set_xticks(())
        ax.set_yticks(())
        ax.set_title(name)
        ax.text(xx.max() - .3, yy.min() + .3, ('%.2f' % score).lstrip('0'),
                size=15, horizontalalignment='right')
        i += 1

figure.subplots_adjust(left=.02, right=.98)
plt.show()
print(X,y)
```
