## Backpropagation(역전파 알고리즘)
- Multi layer perceptron을 학습시키기 위한 방법
  - Output layer에서의 error 변화량을 앞선 layer들로 전파한다는 개념
  - $v_i^{(k)}$ : activation function에 들어가기 전의 값
    - $v_i^{(2)}$의 경우 각 간선 공식에 의해 결정되어짐 $\rightarrow w_{1,1}^{(1)}x_{1}^{(1)}+w_{1,1}^{(1)}x_{2}^{(1)}$
  - 미분을 통하여 접근
  - $w_{i,j}^{(k)}$ : i층에서 j층으로(거꾸로) 가는 k번째 층의 간선
  - $v_{i}^{(k)}$ : 각 간선의 공식을 통해 얻은 값
  

- 단위를 맞춰주는 과정에서의 혼란이 발생하는 경우가 잦다.
  - 아래의 그림에서 $e,v,x$의 단위가 모두 다르듯.
- 여기서 각 값에 대한 **미분을 하게 되면 그 전값으로 돌아간다는 취지**에서 역전파 알고리즘이 시작됨
- 거꾸로 돌아가면서 결론적으로 $w_{1,1}^{(1)}$을 구하기 위해 그 과정을 돌아가고 돌아가며 미분하는 것을 알 수 있음


- (같은 색으로 표시된 공식끼리 단위가 같음)
- **$w_{1,1}^{(1)}$을 구하기 위해서 에러량이 필요함.**
  - 이를 구하기 위해 단위를 바꾸고, 단위 평균을 구하고, 이를 미분하여 에러량을 구하는 과정을 역으로 거치게 된다.
  - 단위를 바꾼다는 것이 엄밀히 말하면 수학에서의 Chain Rule에 속함

- $x_i^{(k)}, v_i^{(k+1)}$사이의 관계
  - k번째 레이어의 노드 수가 $n_k$라 할 때
  - $v_i^{(k+1)}=w_{1,i}^{(k)}x_1^{(k)}+...+w_{n_k}^{(k)}x_{n_k}^{(k)}$
  - **x들로 인해 만들어진 v은 Linear Combination 관계, v와 다음 레이어의 x는 Activation function 관계**

- $v_i^{(k)}, x_i^{(k)}$와의 관계
  - 시그모이드 함수를 activation function으로 사용하면
  

- Weight를 업데이트하는 기본 수식
  - $(w_{i,j}^{(k)})'=w_{i,j}^{(k)}-LR\frac{\delta L}{\delta w_{i,j}^{(k)}}$
    - Squared error를 가정
      - $L = L_1 + L_2$
      - $L_i=\frac{1}{2}(y_i-x_i^{(3)})^2$ 로 정의


- 기본적인 계산의 배경은 **처음 오차가 클수록 점점 커지는 $e_1^{(3)}$을 이해하는 것**이 핵심이다.

- $LR=1$이라 하자. $e_1^{(3)}$을 정의해 간단히 나타내면


- 비슷한 느낌으로


- $(w_{1,1}^2)'=w_{1,1}^{(2)}-\frac{\delta L}{\delta w_{1,1}^{(2)}}=w_{1,1}^{(2)}-e_1^{(3)}x_1^{(2)}$
  - 여기서 $e_1^{(3)}$가 의미하는 것 : 오차값과 activation function의 미분값의 곱
  - $e_1^{(3)}=(x_1^{(3)}-y_1)x_1^{(3)}(1-x_1^{(3)})$
    - 오차가 커질수록 커지는 어떠한 값이라 이해하면 **오차가 클수록 값을 크게 바꾼다는 뜻이 된다.**

- 여기서 Sigmoid 함수의 한계점
  - Sigmoid를 미분하게 되면 정규분포 형태마냥 미분이 된다.
    - 이는 x의 값이 점점 커질수록 그 값은 0에 수렴하게 된다는 것
    - **Weight를 정상적으로 업데이트할 수 없는 문제** - 정보가 사라지는 문제가 발생함

- 앞의 값 $(e_1^{(3)}w_{1,1}^{(2)}+e_2^{(3)}w_{2,1}^{(3)})$의 값이 나오는 이유
- 계산된 값


- 이때의 빨간 weight의 계산은 아래와 같아짐


- **결과의 요약**




- 각각의 식이 도출되는 과정을 곰곰이 살핀다.
  - $(e_1^{(3)}w_{1,1}^{(2)}+e_2^{(3)}w_{2,1}^{(3)})$가 의미하는 것
    - **이후의 error값이 이전의 error에 영향을 주기 때문에 Backpropagation(역전파)라는 말이 나오게 된 것임**
