## K-Nearest Neighborhood

- 주변 관측치에 대한 정보를 주어 새로운 관측치에 대한 분류를 진행하는 알고리즘
- 아래의 초록 점은 어디에 포함해야 할까?

- **K는 어떻게 정하지?**(몇 개의 변수를 가지고 판단을 할 것인가)
  - 너무 큰 k : 미세한 경계부분을 잘못 분류할 것
  - 너무 작은 k : 이상치의 영향을 크게 받을 것. 패턴이 직관적이지 않을 것
  
  
- **Majority voting**
  - Blue가 red에 비해 훨씬 많다면?
  - 거리에 반비례하는 Weight를 줄 필요가 있음
  
  
- **중요한 변수와 불필요한 변수가 섞여 있다면?**
   - 중요한 변수와 불필요한 변수를 선별할 필요가 있음

- 종속변수가
  - **범주형**이라면
    - k-nearest neighbor 중 가장 많이 나타나는 범주로 y를 추정
    - Tie 문제를 막기 위해 k는 홀수로 정하는 것이 좋다.
  - **연속형**이라면
    - knn의 대표값(평균)으로 y를 추정
    - Inverse distance weighted average 고려 가능

- **그럼 거리는 어떻게 구하지?** 설명변수가
  - **범주형** 변수라면
    - Hamming distance = $D_H=\sum_{J}^{j=1}(x_j\neq y_j)$ // 각 코드당 다른것으로 판단
  - **연속형** 변수 $j=1,\dots,j$라면
    - Euclidian distance : $\sqrt{\sum_{j=1}^{J}(x_j-y_j)^2}$
    - Manhattan distance : $\sum_{j=1}^{J}|x_j-y_j|^2$

## 모델의 정의 및 추정
- 점 $(x,y),N$개의 Training 관측치 $(X_i,Y_i),i=1,\dots,N$에 대하여
  - $(X_{(1)},Y_{(1)}),\dots,(X_{(n)},Y_{(n)})$
    - 단, 다음 조건에 따라 정렬되어 있음
    - $d(X_{(1)},x) \leq \dots \leq d(X_{(n)},n)$

- Distance $d(a,b)$의 선택
  - 범주형 변수 : Hamming distance
  - 연속형 변수 : Euclidian distance, Manhattan distance

### 종속 변수가
- **범주형 변수** $m=1,\dots,M$ 라면
  - 근처 k개 중에 가장 많은 범주를 선택
  - $\hat{p}_m=\frac{\sum_{i=1}^{k}(Y_{(i)}=m)}{k}$ : 모든 클래스에 대한 값들을 구함
  - $\hat{p}=argmax_{(m=1,\dots,M)}\hat{p}_m$ : 그 클래스 중에서 가장 높은 값
    
    
- **연속형 변수**라면
  - 근처 k개의 평균을 선택
    - $\hat{y}=\frac{\sum_{i=1}^{k}Y_{(i)}}{k}$
  - Inverse distance weighted average 고려(가중치)
    - $\hat{y}=\frac{\sum_{i=1}^{k}\frac{1}{d(X_{(i)},x)}Y_{(i)}}{k}$

## Cross Validation
- 과적합 문제의 해결, Sample Loss 문제를 해결하기 위해 고안
- Training set을 가장 잘 맞히는 머신은 Test set에서는 동작하지 않을 수 있다.
- Training error는 error를 과소추정하는 성향이 있음

- Bayes Error : 최적의 error 상태
- 10-fold CV : Cross Validation을 적용한 모델

- **Test error를 구하려면 데이터를 나누어야 하므로 데이터의 소실이 발생**
- error를 과소추정하지 않으면서 데이터의 소실을 최소화 하려면?
  - **데이터를 나누어 번갈아 가며 traning과 test set 역할을 한다**

### Cross-validation
- k-fold cross-validation을 구하는 방법
  - ex: $k=5$인 경우

- Cross_validation error
  - $CV(\hat{f})=\frac{1}{N}\sum_{i=1}^{N}L(y_i\hat{f}^{-k(i)}(x_i))$
  - $\hat{f}^{-k(i)}$ : i 관측치가 속한 part를 **제외하고**피팅한 모델

- k의 결정
  - 너무 큰 k  : 미세한 경계부분 분류가 아쉬움
  - 나무 작은 k : 과적합, 이상치 영향을 크게 받음, 패턴이 직관적이지 않음
  - **test error를 기준으로 k를 합리적으로 결정**

- Training error : k=1에서 가장 낮음. 과적합 가능성
- Test error : 데이터에 따라 최적의 k가 존재
- K의 결정 : **Test error를 작게하는 k, Cross-Validation 이용**
  - 아래의 그림에서는 7일 때의 Test error가 제일 작다
  - **10-fold CV가 처음으로 올라는 값**을 k로 쓰는 것이 일반적임

- Y축 정보가 추가됨에 따른 비효율의 예시

- 분포상으로는 초록에 가까운 변수이지만, KNN을 돌렸을 경우, 빨간 변수로 인식하는 경우가 충분히 생긴다.
  - **현재의 y축은 클래스의 정보를 크게 담지 않기 때문이다. 차원 축소를 해도 무방함.**
    - Principal Component Analysis, Canonical Correlation Analysis 등의 방법으로 차원 축소 진행 후 KNN 시행

## KNN 실습
## 1. 데이터, 모듈 불러오기 및 KNN 피팅

```python
from sklearn import neighbors, datasets
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

```python
iris=datasets.load_iris()
X=iris.data[:,:2]
y=iris.target
```

- 모델 구축

```python
clf=neighbors.KNeighborsClassifier(5)
clf.fit(X,y)
```

```python
y_pred=clf.predict(X)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y,y_pred)
```

- 오차가 꽤 심함을 알 수 있다

## 2. Cross-validation을 활용한 최적의 k 찾기
- 함수 불러오기

```python
from sklearn.model_selection import cross_val_score
```

- CV 진행

```python
k_range=range(1,100)
k_score=[]
for k in k_range:
    knn=neighbors.KNeighborsClassifier(k)
    scores=cross_val_score(knn,X,y,cv=10,scoring='accuracy')#정확도 추출
    k_score.append(scores.mean())#정확도의 평균을 저장
```

```python
plt.plot(k_range,k_score)
plt.xlabel('Value of K for KNN')
plt.ylabel('CV accuracy')
plt.show()
```

- 이를 통해 k=45쯤에서 정확도가 제일 높은 것을 알 수 있음

## 추가 : Weight를 준 kNN

```python
n_neighbors = 40

h = .02  # step size in the mesh

cmap_light = ListedColormap(['#FFAAAA', '#AAFFAA', '#AAAAFF'])
cmap_bold = ListedColormap(['#FF0000', '#00FF00', '#0000FF'])

for weights in ['uniform', 'distance']:
    clf = neighbors.KNeighborsClassifier(n_neighbors, weights=weights)
    clf.fit(X, y)

    # Plot the decision boundary. For that, we will assign a color to each
    # point in the mesh [x_min, x_max]x[y_min, y_max].
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
                         np.arange(y_min, y_max, h))
    Z = clf.predict(np.c_[xx.ravel(), yy.ravel()])

    # Put the result into a color plot
    Z = Z.reshape(xx.shape)
    plt.figure()
    plt.pcolormesh(xx, yy, Z, cmap=cmap_light)

    # Plot also the training points
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=cmap_bold,
                edgecolor='k', s=20)
    plt.xlim(xx.min(), xx.max())
    plt.ylim(yy.min(), yy.max())
    plt.title("3-Class classification (k = %i, weights = '%s')"
              % (n_neighbors, weights))

plt.show()
```

- 직관적인 바운더리가 나오는 것도 중요함
  - 여기서는 distance를 준 것이 더 직관적으로 보임

```python
#온전한 데이터 감지를 하는지 검사
np.random.seed(0)
X = np.sort(5 * np.random.rand(40, 1), axis=0)
T = np.linspace(0, 5, 500)[:, np.newaxis]
y = np.sin(X).ravel()
y[::5] += 1 * (0.5 - np.random.rand(8))
```

```python
knn = neighbors.KNeighborsRegressor(n_neighbors)
y_ = knn.fit(X, y).predict(T)
```

```python
n_neighbors = 5

for i, weights in enumerate(['uniform', 'distance']):
    knn = neighbors.KNeighborsRegressor(n_neighbors, weights=weights)
    y_ = knn.fit(X, y).predict(T)

    plt.subplot(2, 1, i + 1)
    plt.scatter(X, y, c='k', label='data')
    plt.plot(T, y_, c='g', label='prediction')
    plt.axis('tight')
    plt.legend()
    plt.title("KNeighborsRegressor (k = %i, weights = '%s')" % (n_neighbors,
                                                                weights))

plt.tight_layout()
plt.show()
```

- distance(가중치)를 줬을 경우, 삐져나온 데이터에 대해서도 감지를 하는 것을 볼 수가 있다.
- **무조건 가중치를 준 것이 더 좋다고 단언할 수는 없다.**
