## 의사결정나무
- 사람들에게 어떤 메시지를 줄 수 있는가? 의사결정나무는 그 능력이 뛰어나다
- 변수들로 기준을 만들고, 이것을 통하여 샘플을 분류하고 분류된 집단의 성질을 통하여 추정하는 모형
  - 장점 : 해석력이 높음. 직관적. 범용성
  - 단점 : 높은 변동성. 샘플에 민감할 수 있음
  

### 각종 용어
- Node : 분류의 기준이 되는 변수가 위치. 이것을 기준으로 샘플을 나눔
  - Parent Node : 상대적인 개념. 상위노드
  - Child Node : 하위노드
  - Root Node : 상위노드가 없는 가장 위의 노드
  - Leaf Node : 하위 노드가 없는 가장 아래의 노드
  - Internal Node : Leaf Node가 아닌 모드
- Edge : 샘플을 분류하는 조건이 위치하는 곳
- Depth : Root Node에서 특정 노드까지 도달하기 위해 거쳐야하는 Egde의 수

- 반응 변수에 따라
  - 범주형 변수라면 **분류 트리**, 연속형 변수라면 **회귀 트리**를 이용한다.
  

## 엔트로피

- 결과가 Yes, No로 두 가지인 경우 2 class prob이라고 함(3개라면 3 이렇게 되겠지)

- 노란 곡물만 나와있는 상황 : $P_+$가 1일때(최저점), 0이라면 검정 곡물만 나와있는 상황임
  - 엔트로피가 0일때의 상황임
- 이 둘이 잘 섞여있는 경우 : 엔트로피가 최대인 상황
- 이와 비슷한 함수 : $-p(p-1)+C$ : 극대값 1인 이차함수 느낌임

### 엔트로피의 정의
- $-plog_2p-(1-p)log_2(1-p)$ : 디게 예쁜 위의 그래프가 나옴
- 직관적 정의 : 0또는 1일 확률이 최소, 0.5일 확률이 최대가 되게 하는 함수

- Tennis 데이터에서의 의미
  - 우리가 가지고 있는 속성을 이용하여 혼합도를 최대로 하는 값에서의 최적값을 찾아내는 것
  

## Information Gain
- Information Gain = $Entropy_{before}-Entropy_{after}$
  - 이를 통해 각각의 변수를 평가하게 됨. 변수의 우선순위(measure)를 주는 기능을 가지고 있음
- Decision Tree의 특정 node 이전과 이후의 Entropy 차이


- Node를 통해 갈라진 3개의 가지 아래의 Information Gain을 구할 때

  - 위의 before에서 3개로 찢어진 after상황에 대해 각각의 Information Gain을 구함
  - 엔트로피 공식에서 각각의 확률만을 구하면 엔트로피를 쉽게 도출해낼 수 있음
    - 사진에서의 Yes 전체 확률은 $9/14$
    - 각각 찢어진 P에 대해서도 각각에 대한 확률을 구하면 됨    

- Node를 통해 분기된 가지에서의 Information Gain 요약이 필요함
  - 여러가지 방법이 있겠지만, 다음과 같은 공식을 활용하여 요약
  - 각각의 엔트로피에 가중치를 주는 형태임
  - 따라서 $Entropy_{after}=\frac{n_1Entropy_{after1}+n_2Entropy_{after2}+n_3Entropy_{after3}}{n_1+n_2+n_3}$
  - $=\frac{n_1}{n_1+n_2+n_3}Entropy_{after1}+\frac{n_2}{n_1+n_2+n_3}Entropy_{after2}+\frac{n_3}{n_1+n_2+n_3}Entropy_{after3}$

#### 자료가 많이 섞여있을수록(초기상태일수록) 엔트로피가 높다. 트리의 Depth가 커질수록 엔트로피는 낮아지는 현상


## Classification Tree의 정의

- Tree 조건에 따라 X가 가질 수 있는 영역을 Block으로 나누는 개념
  - 나누어진 영역안에 속하는 샘플의 특성을 통하여 Y를 추정
- m번째 노드가 k범주일 확률
  - $\hat{p}_{mk}=\frac{1}{N_m}\sum_{x_i \in R_m}I(y_i=k)$
  - 상대비율의 개념

### **$R_m$의 구성**
- 각각의 독립변수에 대하여
  - **범주형** : 각 범주에 따라 / **연속형** : 여러개의 영역으로 임의로 나눔
- 나누어둔 영역들에 대하여, **아래 measure 중 가장 좋은 값으로 만드는(수치가 가장 낮은) 변수와 기준**을 선택하도록 한다.
  
  - 엔트로피하고 지니 인덱스를 많이 쓴다고 함(둥그렇자너)

- 결정된 $R_m$에 대하여
  - $\hat{p}_{mk}=\frac{1}{N_m}\sum_{x_i \in R_m}I(y_i=k)$
  - 추정된 Y의 범주 : $k(m)=argmax_{(k=1,...,K)}\hat{p}_{mk}$
  - **각각의 영역에 대해서 $\hat{p}_{mk}$이 최대가 되는 값들을 각각 구해주는 것이 포인트**
  

## Regression Tree의 정의
- Tree 조건에 따라, X가 가질 수 있는 영역을 Block으로 나누는 개념
  - 나누어진 영역안에 속하는 샘플의 특성을 통하여 Y를 추정
  


- Y의 추정값
  - $f(x)=\sum_{m=1}^{M}c_mI(x\in R_m)$ : $c_m$이 각각의 영역에 실수를 곱하는 형태가 되는 것임(실수 형태)
  

- $R_m$의 구성
  - 아래와 같이 임의로 영역을 나눔 : 적절한 $s$값을 찾는다.
  
  
  
  - 나누어둔 영역들에 대하여 아래 measure를 가장 좋은 값으로 만드는 변수와 기준을 선택
  

- **결정된 $R_m$에 대하여**
  - $\hat{c}_m=avg(y_i|x_i\in R_m)$
  - $\hat{y}=\hat{f}(x)=\sum_{m=1}^{M}\hat{c}_mI(x\in R_m)$
  

#### $R_m$을 결정하는 방법만 clssification과 다르다는 것 외에는 전체적으로 비슷한 구조를 띠는 것을 알 수 있다.

## 2. 의사결정나무 구축 및 시각화

- 트리 구축

```python
clf=tree.DecisionTreeClassifier() #현재 속성이 범주형이기 때문에 Classifier 선택
clf=clf.fit(iris.data,iris.target)
```

- 트리의 시각화

```python
dot_data=tree.export_graphviz(clf,out_file=None,#그림을 직접 그리기 위한 세팅
                              feature_names=iris.feature_names,
                              class_names=iris.target_names,
                              filled=True, rounded=True,#그림에 색상을 넣어줌. 소수점 절삭
                             special_characters=True)#특수문자 옵션 허용
```

```python
graph=graphviz.Source(dot_data)
```

```python
graph
```

- 띠용스? 지니 계수를 쓰네? 엔트로피를 사용토록 해보자

- **엔트로피를 활용한 트리**

```python
clf2=tree.DecisionTreeClassifier(criterion='entropy')#공식을 이렇게 바꾸어주어야 함
clf2=clf2.fit(iris.data,iris.target)
```

```python
dot_data2=tree.export_graphviz(clf2,out_file=None,#그림을 직접 그리기 위한 세팅
                              feature_names=iris.feature_names,
                              class_names=iris.target_names,
                              filled=True, rounded=True,#그림에 색상을 넣어줌. 소수점 절삭
                             special_characters=True)#특수문자 옵션 허용
graph2=graphviz.Source(dot_data2)
```

```python
graph2
```

- 사실 fitting을 이런식으로 하면 안 됨. 트리가 너무 많아! overfitting 문제가 발생하게 됨

    - 프루닝에서의 depth 값을 적절히 구하는 방법 : Cross-Validation

- **프루닝 과정**

```python
clf3=tree.DecisionTreeClassifier(criterion='entropy',max_depth=2)#노드의 깊이 제한(적당한 수준에서 끊어지게 하도록 함)
clf3=clf3.fit(iris.data,iris.target)
```

```python
dot_data3=tree.export_graphviz(clf3,out_file=None,#그림을 직접 그리기 위한 세팅
                              feature_names=iris.feature_names,
                              class_names=iris.target_names,
                              filled=True, rounded=True,#그림에 색상을 넣어줌. 소수점 절삭
                             special_characters=True)#특수문자 옵션 허용
graph3=graphviz.Source(dot_data3)
```

```python
graph3
```

- R에서는 max_depth를 구하는 공식이 내장되어 있다. 파이썬에는 그런거 없대
- **프루닝을 했을 때의 Confusion Matrix 구하기**

```python
confusion_matrix(iris.target,clf2.predict(iris.data))#엔트로피
```

```python
confusion_matrix(iris.target,clf3.predict(iris.data))#엔트로피+프루닝
```

- 프루닝 한 것이 과적합 문제를 해결할 수 있는것이지, 이것이 완전히 나쁜 기능은 아니다.

## 4. Decision Regression Tree

- 모듈 불러오기 및 데이터 생성

```python
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

rng=np.random.RandomState(1)
X=np.sort(5*rng.rand(80,1),axis=0)
y=np.sin(X).ravel()
y[::5]+=3*(0.5-rng.rand(16))
```

```python
regr1=tree.DecisionTreeRegressor(max_depth=2)
regr2=tree.DecisionTreeRegressor(max_depth=5)
```

```python
regr1.fit(X,y)
```

```python
regr2.fit(X,y)
```

```python
X_test=np.arange(0.0,5.0,0.01)[:,np.newaxis]
X_test #5가 될때까지 0.01 단위로 테스트 데이터를 만듦
```

```python
y_1=regr1.predict(X_test)
y_2=regr2.predict(X_test)
```

```python
y_1
```

```python
plt.figure()
plt.scatter(X, y, s=20, edgecolor="black",
            c="darkorange", label="data")
plt.plot(X_test, y_1, color="cornflowerblue",
         label="max_depth=2", linewidth=2)
plt.plot(X_test, y_2, color="yellowgreen", label="max_depth=5", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

- depth가 높아질수록 이상치에 민감하게 반응하는 결과를 초래함
- 간단한 모델이 더 직관적이기 때문에 더 좋은 모델이 아닐까 생각해봄
